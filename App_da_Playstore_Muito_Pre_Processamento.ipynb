#3° Pré-Processamento

```python
# Funções para tratamento do campo "SIZE" (TAMANHO)
# Função que pega somente o último caracter do campo size
df['Size_type'] = df['Size'].str.strip().str[-1]
df['Size_type']
```

```python
### Transformação dos valores 
# lista a ser preenchida no looping
x=[]
# Looping de conversão
for i in df['Size_type']:
  if i=='e':
    x.append(0)
  elif i=='M':
    x.append(1024)
  elif i=='k':
    x.append(1)

# Preenchendo o valor do campo por "x"
df['Size_type']=x

# Extraindo o sinal de "+"
df['size_t']=df.Size.str.extract('(\d+)')

# Substituindo o missing por zero
df['size_t']=df['size_t'].fillna(0)

# multiplicando o campo pela sua dimensão (M: mega, então multiplica por 1024)
df['Size']=df['Size_type'].astype(int)*df['size_t'].astype(int)

# deleta
del df['Size_type']
del df['size_t']

#imputação de missing pela media
df['Size']=df['Size'].replace(0,df['Size'].mean())
```

```python
# Substituindo free por "0"
df['Price']=df['Price'].replace('Free',0).astype(float)
```

```python
# Transformando o campo de data em datetime
df['Last_Updated_On'] = pd.to_datetime(df['Last_Updated_On'])
```

```python
# Criando campos temporais
k=['Last_Updated_On']
for i in k:
  df[i+'_'+'year'] = df[i].dt.year
  df[i+'_'+'day'] = df[i].dt.day
  df[i+'_'+'weekofyear'] = df[i].dt.weekofyear
  df[i+'_'+'month'] = df[i].dt.month
  df[i+'_'+'dayofweek'] =df[i].dt.dayofweek
  df[i+'_'+'weekend'] = (df[i].dt.weekday >=5).astype(int)
del df['Last_Updated_On']
```

```python
# o train será o dataframe que possui valores no atributo Downloads
train = df[df['Downloads'].isnull()==False]
# o test será o dataframe que não possui valores no atributo Downloads
test = df[df['Downloads'].isnull()==True]
# já que test possui valores vazios no atributo Download, então apago esse atributo
del test['Downloads']
```

```python
# Preparando o TARGET "Downloads"
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['Downloads'] = le.fit_transform(train['Downloads'])
le.classes_

is_hol = train['Downloads'] == 11
df_try = train[is_hol]
train=train.append([df_try],ignore_index=True)
train_df=train
test_df=test
```

```python
train['Downloads']
```

```python
# Separando as Variáveis Explicativas e Target
X = train_df.drop(labels=['Downloads'], axis=1)
y = train_df['Downloads'].values
```

```python
# Realizando a Amostragem
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=101, stratify=y)
```

```python
# Verificando dimensões das bases de dados
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

# 4° e 5° Construção e Avaliação da Máquina Preditiva

![](https://cienciadosdados.com/images/en2.png)

```python
# Importando pacotes de métricas e do algoritmo 
from sklearn.metrics import log_loss
import lightgbm as lgb
```

```python
# Preparando os dados na Máquina Preditiva 
train_data = lgb.Dataset(X_train, label=y_train)
test_data = lgb.Dataset(X_test, label=y_test)

```

```python
# Tunagem de Hiperparâmetros
param = {
         'num_iterations': 400,
         'metric': 'multi_logloss',
         'objective': 'multiclass',
         'num_class': 18,
         'learning_rate': 0.09,
         'boosting': 'dart',  
         'num_leaves': 31,
         'max_depth': -1,
         'min_data_in_leaf': 30,
         'bagging_fraction':0.9,
         'bagging_freq': 2,
         'feature_fraction': 0.9,
         'lambda_l2': 0.9,
         'min_data_per_group': 500,
         'max_bin': 255,
         'is_unbalance':True
         
         }

```

```python
# Criando a Máquina Preditiva - Treinando e Testando 
clf = lgb.train(params=param, 
                early_stopping_rounds=200,
                verbose_eval=100,
                train_set=train_data,
                valid_sets=[test_data])

y_pred = clf.predict(X_test)
```
