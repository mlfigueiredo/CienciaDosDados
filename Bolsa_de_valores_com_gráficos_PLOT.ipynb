```python
import plotly.graph_objs as go
import matplotlib.pyplot as plt

Close= [go.Scatter(
                  x=df.Date, 
                  y=df.Close, 
                  name= 'Petr4 Close', 
                  line= dict(color='#000000'), 
                  opacity=0.8
                  )
        ]
data=[Close]
layout= dict(
    title='Faixa de Data Customizada',  
    title_x=0.5,
    x_axis=dict(
        range=['2017-01-01', '2017-12-31'])
    )
fig= dict(data=data, layout=layout)
py.iplot(Close)
```

```python
trace= go.Candlestick(x=df.Date,
                      open=df.Open,
                      high=df.High,
                      low=df.Low,
                      close=df.Close)
data = [trace]
py.iplot(data, filename='cladlestick')
```

```python
trace= go.Candlestick(x=df.Date,
                      open=df.Open,
                      high=df.High,
                      low=df.Low,
                      close=df.Close)
layout= go.Layout(
    title='pre√ßos em candlesticks',
    title_x=0.5,
    xaxis= dict(
        rangeslider= dict(
            visible=False
        )
    )
)
data = [trace]
fig= go.Figure(data=data, layout=layout)
py.iplot(fig)
```

```python
#Trabalhando com m√©dias
#Aqui n√£o √© m√©dia movel, pois os dados n√£o ser√£o movidos, sen√£o teria que criar um la√ßo
#m√©dia de 9 dias
df['MM_9']= df.Close.rolling(window=9).mean()
#m√©dia de 17 dias
df['MM_17']= df.Close.rolling(window=17).mean()
```

```python
df.head(20)
```

```python
close= go.Scatter(
    x=df.Date,
    y=df.Close,
    name='PETR4 Close',
    line= dict(color= '#330000'),
    opacity= 0.8
)
MM_9= go.Scatter(
    x=df.Date,
    y=df.MM_9,
    name='M√©dia m√≥vel 9 per√≠odos',
    line= dict(color= '#FF8000'),
    opacity= 0.8
)
MM_17= go.Scatter(
    x=df.Date,
    y=df.MM_17,
    name='M√©dia m√≥vel 17 per√≠odos',
    line= dict(color= '#FF8000'),
    opacity= 0.8
)
data=[close, MM_9, MM_17]
py.iplot(data)
```

```python
#m√©dia de 17 dias
df['MM_3']= df.Close.rolling(window=3).mean()
```

```python
close= go.Scatter(
    x=df.Date,
    y=df.Close,
    name='PETR4 Close',
    line= dict(color= '#330000'),
    opacity= 0.8
)
MM_3= go.Scatter(
    x=df.Date,
    y=df.MM_3,
    name='M√©dia m√≥vel 3 per√≠odos',
    line= dict(color= '#AF8F00'),
    opacity= 0.8
)
MM_9= go.Scatter(
    x=df.Date,
    y=df.MM_9,
    name='M√©dia m√≥vel 9 per√≠odos',
    line= dict(color= '#FF8000'),
    opacity= 0.8
)
MM_17= go.Scatter(
    x=df.Date,
    y=df.MM_17,
    name='M√©dia m√≥vel 17 per√≠odos',
    line= dict(color= '#FF8000'),
    opacity= 0.8
)
data=[close, MM_3, MM_9, MM_17]
py.iplot(data)
```
