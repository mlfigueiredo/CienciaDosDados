#Iniciando o **Pywedge_Charts** 




Cria diferentes tipos de gráficos de visualização de Dados de forma interativa, ou seja, sem a necessidade de configurações adicionais:

* Gráfico de dispersão
* Gráfico de pizza
* Bar Plot
* Violino
* Box Plot
* Gráfico de Distribuição
* Histograma
* Gráfico de Correlação



## O "Pywedge_Charts" precisa dos seguintes parâmetros:


**1° Dataframe**

**2° c** = Eliminar colunas redundantes (Id por exemplo)

**3° y** = A variável Target (classe, etc)


###Case 1

```
#Comando de Instalação
!pip install pywedge
```

```
#Importação de Bibliotecas
import pywedge as pw
```

```
#importação da fonte de dados
import pandas as pd
train = pd.read_csv('/content/risco.csv')
```

```
train.head()
```

```
#Input de Parâmetros
mc = pw.Pywedge_Charts(train, c='id_cliente', y = 'Risco')
```

```
#Run
charts = mc.make_charts()
```
