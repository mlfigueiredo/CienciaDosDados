### Extração de E-mail de um texto

```
#Biblioteca Regular Expressions
import re
# Texto para buscar o E-mail
text = "Observa-se que o correio eletrônico, \
edau@gmail.com, deixa de ser apenas um meio de troca de \
mensagens entre pessoas para se tornar um grande fator na \
produtividade das empresas. Grandes empresas como visaofin@gmail.com , estão, cada vez \
mais, usando o correio eletrônico para desempenhar papéis decisivos\
 em suas negociações. Dentro disso, o e-mail suporte@cienciadosdados.com, uma intranet pode ser estabelecida \
 para tornar a comunicação de funcionários com outros grupos, tornando \
 assim mais fácil o trabalho e eliminando mensagens em massa e outras mensagens indesejada. email para contato marcosleandrofigueiredo@gmail.com."

re.findall(r"[\w.-]+@[\w.-]+", text)
```

```
#Botão direito - Exibir código fonte da Página
html = """ 

<!DOCTYPE html>
<html class="client-nojs" lang="pt" dir="ltr">
<head>
<meta charset="UTF-8"/>
<title>Colesterol – Wikipédia, a enciclopédia livre</title>
<script>document.documentElement.className="client-js";RLCONF={"wgBreakFrames":false,"wgSeparatorTransformTable":[",\t."," \t,"],"wgDigitTransformTable":["",""],"wgDefaultDateFormat":"dmy","wgMonthNames":["","janeiro","fevereiro","março","abril","maio","junho","julho","agosto","setembro","outubro","novembro","dezembro"],"wgRequestId":"0839aa36-3afa-41b2-8e50-3d1e8fa1f603","wgCSPNonce":false,"wgCanonicalNamespace":"","wgCanonicalSpecialPageName":false,"wgNamespaceNumber":0,"wgPageName":"Colesterol","wgTitle":"Colesterol","wgCurRevisionId":61241919,"wgRevisionId":61241919,"wgArticleId":94201,"wgIsArticle":true,"wgIsRedirect":false,"wgAction":"view","wgUserName":null,"wgUserGroups":["*"],"wgCategories":["!CS1 manut: Nomes múltiplos: lista de autores","!Páginas que usam hiperligações mágicas PMID","!CS1 inglês-fontes em língua (en)","Lípidos","Esteróis"],"wgPageContentLanguage":"pt","wgPageContentModel":"wikitext","wgRelevantPageName":"Colesterol","wgRelevantArticleId":94201,
"wgIsProbablyEditable":true,"wgRelevantPageIsProbablyEditable":true,"wgRestrictionEdit":[],"wgRestrictionMove":[],"wgVisualEditor":{"pageLanguageCode":"pt","pageLanguageDir":"ltr","pageVariantFallbacks":"pt"},"wgMFDisplayWikibaseDescriptions":{"search":true,"nearby":true,"watchlist":true,"tagline":true},"wgWMESchemaEditAttemptStepOversample":false,"wgWMEPageLength":30000,"wgNoticeProject":"wikipedia","wgMediaViewerOnClick":true,"wgMediaViewerEnabledByDefault":true,"wgPopupsFlags":10,"wgULSCurrentAutonym":"português","wgEditSubmitButtonLabelPublish":true,"wgCentralAuthMobileDomain":false,"wgULSPosition":"interlanguage","wgULSisCompactLinksEnabled":true,"wgWikibaseItemId":"Q43656","GEHomepageSuggestedEditsEnableTopics":true,"wgGETopicsMatchModeEnabled":true,"wgGEStructuredTaskRejectionReasonTextInputEnabled":false,"wgSiteNoticeId":"2.30"};RLSTATE={"skins.vector.user.styles":"ready","ext.gadget.FeedbackHighlight-base":"ready","ext.globalCssJs.user.styles":"ready","site.styles":"ready",
"user.styles":"ready","skins.vector.user":"ready","ext.globalCssJs.user":"ready","user":"ready","user.options":"loading","ext.cite.styles":"ready","mediawiki.page.gallery.styles":"ready","mediawiki.ui.button":"ready","skins.vector.styles":"ready","skins.vector.icons":"ready","mediawiki.ui.icon":"ready","ext.visualEditor.desktopArticleTarget.noscript":"ready","ext.wikimediaBadges":"ready","ext.uls.interlanguage":"ready","wikibase.client.init":"ready","ext.dismissableSiteNotice.styles":"ready"};RLPAGEMODULES=["ext.cite.ux-enhancements","site","mediawiki.page.ready","mediawiki.toc","skins.vector.js","skins.vector.es6","mmv.head","mmv.bootstrap.autostart","ext.visualEditor.desktopArticleTarget.init","ext.visualEditor.targetLoader","ext.eventLogging","ext.wikimediaEvents","ext.navigationTiming","ext.cx.eventlogging.campaigns","ext.imagesuggestions.notifications.styles","ext.centralNotice.geoIP","ext.centralNotice.startUp","ext.gadget.Topicon","ext.gadget.Metacaixa","ext.gadget.TitleRewrite"
,"ext.gadget.ElementosOcultaveis","ext.gadget.FeedbackHighlight","ext.gadget.ReferenceTooltips","ext.gadget.NewVillagePump","ext.gadget.wikibugs","ext.gadget.charinsert","ext.gadget.requestForAdminship","ext.gadget.EnhancedSearch","ext.gadget.WikiMiniAtlas","ext.gadget.PagesForDeletion","ext.centralauth.centralautologin","ext.popups","ext.uls.compactlinks","ext.uls.interface","ext.cx.uls.quick.actions","ext.growthExperiments.SuggestedEditSession","ext.dismissableSiteNotice"];</script>
<script>(RLQ=window.RLQ||[]).push(function(){mw.loader.implement("user.options@12s5i",function($,jQuery,require,module){mw.user.tokens.set({"patrolToken":"+\\","watchToken":"+\\","csrfToken":"+\\"});});});</script>
<link rel="stylesheet" href="/w/load.php?lang=pt&amp;modules=ext.cite.styles%7Cext.dismissableSiteNotice.styles%7Cext.uls.interlanguage%7Cext.visualEditor.desktopArticleTarget.noscript%7Cext.wikimediaBadges%7Cmediawiki.page.gallery.styles%7Cmediawiki.ui.button%2Cicon%7Cskins.vector.icons%2Cstyles%7Cwikibase.client.init&amp;only=styles&amp;skin=vector-2022"/>
<script async="" src="/w/load.php?lang=pt&amp;modules=startup&amp;only=scripts&amp;raw=1&amp;skin=vector-2022"></script>
<meta name="ResourceLoaderDynamicStyles" content=""/>
<link rel="stylesheet" href="/w/load.php?lang=pt&amp;modules=ext.gadget.FeedbackHighlight-base&amp;only=styles&amp;skin=vector-2022"/>
<link rel="stylesheet" href="/w/load.php?lang=pt&amp;modules=site.styles&amp;only=styles&amp;skin=vector-2022"/>
<meta name="generator" content="MediaWiki 1.39.0-wmf.23"/>
<meta name="referrer" content="origin"/>
<meta name="referrer" content="origin-when-crossorigin"/>
<meta name="referrer" content="origin-when-cross-origin"/>
<meta name="format-detection" content="telephone=no"/>
<meta property="og:image" content="https://upload.wikimedia.org/wikipedia/commons/thumb/9/9a/Cholesterol.svg/1200px-Cholesterol.svg.png"/>
<meta property="og:image:width" content="1200"/>
<meta property="og:image:height" content="807"/>
<meta property="og:image" content="https://upload.wikimedia.org/wikipedia/commons/thumb/9/9a/Cholesterol.svg/800px-Cholesterol.svg.png"/>
<meta property="og:image:width" content="800"/>
<meta property="og:image:height" content="538"/>
<meta property="og:image" content="https://upload.wikimedia.org/wikipedia/commons/thumb/9/9a/Cholesterol.svg/640px-Cholesterol.svg.png"/>
<meta property="og:image:width" content="640"/>
<meta property="og:image:height" content="431"/>
<meta name="viewport" content="width=1000, initial-scale=1.0"/>
<meta property="og:title" content="Colesterol – Wikipédia, a enciclopédia livre"/>
<meta property="og:type" content="website"/>
<link rel="preconnect" href="//upload.wikimedia.org"/>
<link rel="alternate" media="only screen and (max-width: 720px)" href="//pt.m.wikipedia.org/wiki/Colesterol"/>
<link rel="alternate" type="application/x-wiki" title="Editar" href="/w/index.php?title=Colesterol&amp;action=edit"/>
<link rel="apple-touch-icon" href="/static/apple-touch/wikipedia.png"/>
<link rel="shortcut icon" href="/static/favicon/wikipedia.ico"/>
<link rel="search" type="application/opensearchdescription+xml" href="/w/opensearch_desc.php" title="Wikipédia (pt)"/>
<link rel="EditURI" type="application/rsd+xml" href="//pt.wikipedia.org/w/api.php?action=rsd"/>
<link rel="license" href="https://creativecommons.org/licenses/by-sa/3.0/"/>
<link rel="canonical" href="https://pt.wikipedia.org/wiki/Colesterol"/>
<link rel="dns-prefetch" href="//meta.wikimedia.org" />
<link rel="dns-prefetch" href="//login.wikimedia.org"/>
</head>
<body class="skin-vector skin-vector-search-vue mediawiki ltr sitedir-ltr mw-hide-empty-elt ns-0 ns-subject mw-editable page-Colesterol rootpage-Colesterol skin-vector-2022 action-view"><div class="mw-page-container vector-layout-grid vector-toc-visible">
	<span id="top-page"></span>
	<a class="mw-jump-link" href="#content">Saltar para o conteúdo</a>
	<div class="mw-page-container-inner  vector-toc-visible">

<input
	type="checkbox"
	id="mw-sidebar-checkbox"
	class="mw-checkbox-hack-checkbox"
	>

<header class="mw-header">
	<div class="mw-header-aside">
	<label
		id="mw-sidebar-button"
		class="mw-checkbox-hack-button mw-ui-icon mw-ui-button mw-ui-quiet mw-ui-icon-element"
		for="mw-sidebar-checkbox"
		role="button"
		aria-controls="mw-panel"
		data-event-name="ui.sidebar"
		tabindex="0"
		title="Menu principal">
		<span>Alternar barra lateral</span>
	</label>
	
<a href="/wiki/Wikip%C3%A9dia:P%C3%A1gina_principal" class="mw-logo">
	<img class="mw-logo-icon" src="/static/images/mobile/copyright/wikipedia.png" alt=""
		aria-hidden="true" height="50" width="50">
	<span class="mw-logo-container">
		<img class="mw-logo-wordmark" alt="Wikipédia"
			src="/static/images/mobile/copyright/wikipedia-wordmark-fr.svg" style="width: 7.4375em; height: 1.125em;">
		<img class="mw-logo-tagline"
			alt=""
			src="/static/images/mobile/copyright/wikipedia-tagline-pt.svg" width="113" height="13" style="width: 7.0625em; height: 0.8125em;">
	</span>
</a>

	</div>
	<div class="mw-header-content">
	
<div id="p-search" role="search" class="vector-search-box-vue  vector-search-box-collapses  vector-search-box-show-thumbnail vector-search-box-auto-expand-width vector-search-box">
	<a href="/wiki/Especial:Pesquisar"
	
		
		
		
		title="Pesquisar na Wikipédia [f]"
		accesskey="f"
		class="mw-ui-button mw-ui-quiet mw-ui-icon mw-ui-icon-element mw-ui-icon-wikimedia-search search-toggle">
		<span>Busca</span>
	</a>
	
	<div>
		<form action="/w/index.php" id="searchform"
			class="vector-search-box-form">
			<div id="simpleSearch"
				class="vector-search-box-inner"
				 data-search-loc="header-moved">
				<input class="vector-search-box-input"
					 type="search" name="search" placeholder="Pesquisar na Wikipédia" aria-label="Pesquisar na Wikipédia" autocapitalize="sentences" title="Pesquisar na Wikipédia [f]" accesskey="f" id="searchInput"
				>
				<input type="hidden" name="title" value="Especial:Pesquisar">
				<input id="mw-searchButton"
					 class="searchButton mw-fallbackSearchButton" type="submit" name="fulltext" title="Procurar páginas que contêm este texto" value="Pesquisar">
				<input id="searchButton"
					 class="searchButton" type="submit" name="go" title="Ir para uma página com este nome exacto, caso exista" value="Ir">
			</div>
		</form>
	</div>
</div>

	<nav class="vector-user-links" aria-label="Ferramentas pessoais" role="navigation">
	

<div id="p-personal-more" class="vector-menu mw-portlet mw-portlet-personal-more vector-user-menu-more"  >
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list">
	<li id="p-createaccount" class="user-links-collapsible-item">
		<a href="/w/index.php?title=Especial:Criar_conta&amp;returnto=Colesterol" class="user-links-collapsible-item mw-ui-button mw-ui-quiet" title="É encorajado a criar uma conta e iniciar sessão; no entanto, não é obrigatório"><span>Criar uma conta</span></a>
	</li>
</ul>
		
	</div>
</div>

	

<div id="p-personal" class="vector-menu mw-portlet mw-portlet-personal vector-user-menu vector-user-menu-logged-out vector-menu-dropdown"  title="Mais opções" >
	<input type="checkbox"
		id="p-personal-checkbox"
		role="button"
		aria-haspopup="true"
		data-event-name="ui.dropdown-p-personal"
		class="vector-menu-checkbox"
		
	/>
	<label
		id="p-personal-label"
		for="p-personal-checkbox"
		class="vector-menu-heading  mw-ui-button mw-ui-quiet mw-ui-icon mw-ui-icon-element mw-ui-icon-wikimedia-ellipsis"
	>
		<span class="vector-menu-heading-label">Ferramentas pessoais</span>
	</label>
	<div class="vector-menu-content">
		<div class="vector-user-menu-create-account"><a href="/w/index.php?title=Especial:Criar_conta&amp;returnto=Colesterol" class="vector-menu-content-item user-links-collapsible-item" title="É encorajado a criar uma conta e iniciar sessão; no entanto, não é obrigatório"><span class="mw-ui-icon mw-ui-icon-userAdd mw-ui-icon-wikimedia-userAdd"></span> <span>Criar uma conta</span></a></div>
<div class="vector-user-menu-login"><a href="/w/index.php?title=Especial:Entrar&amp;returnto=Colesterol" class="vector-menu-content-item vector-menu-content-item-login" title="Aconselhamos-lhe a criar uma conta na Wikipédia, embora tal não seja obrigatório. [o]" accesskey="o"><span class="mw-ui-icon mw-ui-icon-logIn mw-ui-icon-wikimedia-logIn"></span> <span>Entrar</span></a></div>
<div class="vector-user-menu-anon-editor">
	<p>
		Páginas para editores conectados <a href="/wiki/Wikip%C3%A9dia:Boas-vindas" aria-label="Saiba mais sobre edição"><span>saiba mais</span></a>
	</p>
</div>

		<ul class="vector-menu-content-list"><li id="pt-anontalk" class="mw-list-item"><a href="/wiki/Especial:Minha_discuss%C3%A3o" title="Discussão sobre edições feitas a partir deste endereço IP [n]" accesskey="n"><span>Discussão</span></a></li><li id="pt-anoncontribs" class="mw-list-item"><a href="/wiki/Especial:Minhas_contribui%C3%A7%C3%B5es" title="Uma lista de edições feitas a partir deste endereço IP [y]" accesskey="y"><span>Contribuições</span></a></li></ul>
		
	</div>
</div>

</nav>

	</div>
</header>

<div class="mw-workspace-container vector-sidebar-container ">
	
	<div id="mw-navigation">
		

<nav id="mw-panel" class="mw-sidebar" aria-label="Site" role="navigation">
	

<div id="p-navigation" class="vector-menu mw-portlet mw-portlet-navigation vector-menu-portal portal"  >
	<p
		id="p-navigation-label"
		class="vector-menu-heading "
		
	>
		<span class="vector-menu-heading-label">Navegação</span>
	</p>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li id="n-mainpage-description" class="mw-list-item"><a href="/wiki/Wikip%C3%A9dia:P%C3%A1gina_principal" title="Visitar a página principal [z]" accesskey="z"><span>Página principal</span></a></li><li id="n-featuredcontent" class="mw-list-item"><a href="/wiki/Portal:Conte%C3%BAdo_destacado"><span>Conteúdo destacado</span></a></li><li id="n-currentevents" class="mw-list-item"><a href="/wiki/Portal:Eventos_atuais" title="Informação temática sobre eventos atuais"><span>Eventos atuais</span></a></li><li id="n-villagepump" class="mw-list-item"><a href="/wiki/Wikip%C3%A9dia:Esplanada"><span>Esplanada</span></a></li><li id="n-randompage" class="mw-list-item"><a href="/wiki/Especial:Aleat%C3%B3ria" title="Carregar página aleatória [x]" accesskey="x"><span>Página aleatória</span></a></li><li id="n-portals" class="mw-list-item"><a href="/wiki/Portal:%C3%8Dndice"><span>Portais</span></a></li><li id="n-bug_in_article" class="mw-list-item"><a href="/wiki/Wikip%C3%A9dia:Informe_um_erro"><span>Informar um erro</span></a></li></ul>
		
	</div>
</div>

	
	

<div id="p-interaction" class="vector-menu mw-portlet mw-portlet-interaction vector-menu-portal portal"  >
	<p
		id="p-interaction-label"
		class="vector-menu-heading "
		
	>
		<span class="vector-menu-heading-label">Colaboração</span>
	</p>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li id="n-welcome" class="mw-list-item"><a href="/wiki/Wikip%C3%A9dia:Boas-vindas"><span>Boas-vindas</span></a></li><li id="n-help" class="mw-list-item"><a href="/wiki/Ajuda:P%C3%A1gina_principal" title="Um local reservado para auxílio."><span>Ajuda</span></a></li><li id="n-testpage" class="mw-list-item"><a href="/wiki/Ajuda:P%C3%A1gina_de_testes"><span>Página de testes</span></a></li><li id="n-portal" class="mw-list-item"><a href="/wiki/Wikip%C3%A9dia:Portal_comunit%C3%A1rio" title="Sobre o projeto"><span>Portal comunitário</span></a></li><li id="n-recentchanges" class="mw-list-item"><a href="/wiki/Especial:Mudan%C3%A7as_recentes" title="Uma lista de mudanças recentes nesta wiki [r]" accesskey="r"><span>Mudanças recentes</span></a></li><li id="n-maintenance" class="mw-list-item"><a href="/wiki/Wikip%C3%A9dia:Manuten%C3%A7%C3%A3o"><span>Manutenção</span></a></li><li id="n-createpage" class="mw-list-item"><a href="/wiki/Ajuda:Guia_de_edi%C3%A7%C3%A3o/Como_come%C3%A7ar_uma_p%C3%A1gina"><span>Criar página</span></a></li><li id="n-newpages-description" class="mw-list-item"><a href="/wiki/Especial:P%C3%A1ginas_novas"><span>Páginas novas</span></a></li><li id="n-contact-description" class="mw-list-item"><a href="/wiki/Wikip%C3%A9dia:Contato"><span>Contato</span></a></li><li id="n-sitesupport" class="mw-list-item"><a href="//donate.wikimedia.org/wiki/Special:FundraiserRedirector?utm_source=donate&amp;utm_medium=sidebar&amp;utm_campaign=20120521SB001&amp;uselang=pt" title="Ajude-nos"><span>Donativos</span></a></li></ul>
		
	</div>
</div>


<div id="p-tb" class="vector-menu mw-portlet mw-portlet-tb vector-menu-portal portal"  >
	<p
		id="p-tb-label"
		class="vector-menu-heading "
		
	>
		<span class="vector-menu-heading-label">Ferramentas</span>
	</p>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li id="t-whatlinkshere" class="mw-list-item"><a href="/wiki/Especial:P%C3%A1ginas_afluentes/Colesterol" title="Lista de todas as páginas que contêm hiperligações para esta [j]" accesskey="j"><span>Páginas afluentes</span></a></li><li id="t-recentchangeslinked" class="mw-list-item"><a href="/wiki/Especial:Altera%C3%A7%C3%B5es_relacionadas/Colesterol" rel="nofollow" title="Mudanças recentes nas páginas para as quais esta contém hiperligações [k]" accesskey="k"><span>Alterações relacionadas</span></a></li><li id="t-upload" class="mw-list-item"><a href="/wiki/Wikipedia:Carregar_ficheiro" title="Carregar ficheiros [u]" accesskey="u"><span>Carregar ficheiro</span></a></li><li id="t-specialpages" class="mw-list-item"><a href="/wiki/Especial:P%C3%A1ginas_especiais" title="Lista de páginas especiais [q]" accesskey="q"><span>Páginas especiais</span></a></li><li id="t-permalink" class="mw-list-item"><a href="/w/index.php?title=Colesterol&amp;oldid=61241919" title="Hiperligação permanente para esta revisão da página"><span>Hiperligação permanente</span></a></li><li id="t-info" class="mw-list-item"><a href="/w/index.php?title=Colesterol&amp;action=info" title="Mais informações sobre esta página"><span>Informações da página</span></a></li><li id="t-cite" class="mw-list-item"><a href="/w/index.php?title=Especial:Citar&amp;page=Colesterol&amp;id=61241919&amp;wpFormIdentifier=titleform" title="Informação sobre como citar esta página"><span>Citar esta página</span></a></li><li id="t-wikibase" class="mw-list-item"><a href="https://www.wikidata.org/wiki/Special:EntityPage/Q43656" title="Hiperligação para o elemento do repositório de dados [g]" accesskey="g"><span>Elemento Wikidata</span></a></li></ul>
		
	</div>
</div>


<div id="p-coll-print_export" class="vector-menu mw-portlet mw-portlet-coll-print_export vector-menu-portal portal"  >
	<p
		id="p-coll-print_export-label"
		class="vector-menu-heading "
		
	>
		<span class="vector-menu-heading-label">Imprimir/exportar</span>
	</p>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li id="coll-create_a_book" class="mw-list-item"><a href="/w/index.php?title=Especial:Livro&amp;bookcmd=book_creator&amp;referer=Colesterol"><span>Criar um livro</span></a></li><li id="coll-download-as-rl" class="mw-list-item"><a href="/w/index.php?title=Especial:DownloadAsPdf&amp;page=Colesterol&amp;action=show-download-screen"><span>Descarregar como PDF</span></a></li><li id="t-print" class="mw-list-item"><a href="/w/index.php?title=Colesterol&amp;printable=yes" title="Versão para impressão desta página [p]" accesskey="p"><span>Versão para impressão</span></a></li></ul>
		
	</div>
</div>


<div id="p-wikibase-otherprojects" class="vector-menu mw-portlet mw-portlet-wikibase-otherprojects vector-menu-portal portal"  >
	<p
		id="p-wikibase-otherprojects-label"
		class="vector-menu-heading "
		
	>
		<span class="vector-menu-heading-label">Noutros projetos</span>
	</p>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li class="wb-otherproject-link wb-otherproject-commons mw-list-item"><a href="https://commons.wikimedia.org/wiki/Category:Cholesterol" hreflang="en"><span>Wikimedia Commons</span></a></li></ul>
		
	</div>
</div>

	
<div class="mw-sidebar-action">
	<div class="mw-sidebar-action-item">
		<h3 class="mw-sidebar-action-heading">Idiomas</h3>
		<div class="mw-sidebar-action-content">
		<div class="mw-message-box-notice vector-language-sidebar-alert mw-message-box">Nesta Wikipédia, os atalhos de idioma estão na <a href="#p-lang-btn">parte superior da página, em frente ao título do artigo</a>.</div>
		</div>
	</div>
</div>

</nav>

	</div>
</div>
<input type="checkbox" id="vector-toc-collapsed-checkbox" class="mw-checkbox-hack-checkbox">
<div class="mw-table-of-contents-container mw-sticky-header-element">
	<nav id="mw-panel-toc" class="sidebar-toc" role="navigation" aria-labelledby="sidebar-toc-label" data-event-name="ui.sidebar-toc">
	<div id="sidebar-toc-label" class="sidebar-toc-header">
		<p class="sidebar-toc-title">
		Conteúdo
		<button class="vector-toc-uncollapse-button">
			move to sidebar
		</button>
		<button class="vector-toc-collapse-button">
			Esconder
		</button>
		</p>
	</div>
	<ul class="sidebar-toc-contents">
		<li id="toc-mw-content-text"
			class="sidebar-toc-list-item sidebar-toc-level-1">
			<a href="#top-page" class="sidebar-toc-link">
				<div class="sidebar-toc-text">Início</div>
			</a>
		</li>
		<li id="toc-Fórmula_e_estrutura_química"
			class="sidebar-toc-list-item sidebar-toc-level-1 sidebar-toc-list-item-expanded">
			<a class="sidebar-toc-link" href="#Fórmula_e_estrutura_química">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">1</span>Fórmula e estrutura química</div>
			</a>
			
			<ul id="toc-Fórmula_e_estrutura_química-sublist" class="sidebar-toc-list">
			</ul>
		</li>
		<li id="toc-Fisiologia"
			class="sidebar-toc-list-item sidebar-toc-level-1 sidebar-toc-list-item-expanded">
			<a class="sidebar-toc-link" href="#Fisiologia">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">2</span>Fisiologia</div>
			</a>
			
				<button aria-controls="toc-Fisiologia-sublist" class="mw-ui-icon mw-ui-icon-wikimedia-expand mw-ui-icon-small sidebar-toc-toggle">
					Toggle Fisiologia subsection
				</button>
			
			<ul id="toc-Fisiologia-sublist" class="sidebar-toc-list">
				<li id="toc-Função"
			class="sidebar-toc-list-item sidebar-toc-level-2">
			<a class="sidebar-toc-link" href="#Função">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">2.1</span>Função</div>
			</a>
			
			<ul id="toc-Função-sublist" class="sidebar-toc-list">
			</ul>
		</li>
		<li id="toc-Síntese_e_ingestão"
			class="sidebar-toc-list-item sidebar-toc-level-2">
			<a class="sidebar-toc-link" href="#Síntese_e_ingestão">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">2.2</span>Síntese e ingestão</div>
			</a>
			
			<ul id="toc-Síntese_e_ingestão-sublist" class="sidebar-toc-list">
			</ul>
		</li>
		<li id="toc-Regulação"
			class="sidebar-toc-list-item sidebar-toc-level-2">
			<a class="sidebar-toc-link" href="#Regulação">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">2.3</span>Regulação</div>
			</a>
			
			<ul id="toc-Regulação-sublist" class="sidebar-toc-list">
			</ul>
		</li>
		<li id="toc-Excreção"
			class="sidebar-toc-list-item sidebar-toc-level-2">
			<a class="sidebar-toc-link" href="#Excreção">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">2.4</span>Excreção</div>
			</a>
			
			<ul id="toc-Excreção-sublist" class="sidebar-toc-list">
			</ul>
		</li>
		<li id="toc-Fluidos_corporais"
			class="sidebar-toc-list-item sidebar-toc-level-2">
			<a class="sidebar-toc-link" href="#Fluidos_corporais">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">2.5</span>Fluidos corporais</div>
			</a>
			
			<ul id="toc-Fluidos_corporais-sublist" class="sidebar-toc-list">
			</ul>
		</li>
	</ul>
		</li>
		<li id="toc-Importância_clínica"
			class="sidebar-toc-list-item sidebar-toc-level-1 sidebar-toc-list-item-expanded">
			<a class="sidebar-toc-link" href="#Importância_clínica">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">3</span>Importância clínica</div>
			</a>
			
				<button aria-controls="toc-Importância_clínica-sublist" class="mw-ui-icon mw-ui-icon-wikimedia-expand mw-ui-icon-small sidebar-toc-toggle">
					Toggle Importância clínica subsection
				</button>
			
			<ul id="toc-Importância_clínica-sublist" class="sidebar-toc-list">
				<li id="toc-Hipercolesterolemia"
			class="sidebar-toc-list-item sidebar-toc-level-2">
			<a class="sidebar-toc-link" href="#Hipercolesterolemia">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">3.1</span>Hipercolesterolemia</div>
			</a>
			
			<ul id="toc-Hipercolesterolemia-sublist" class="sidebar-toc-list">
			</ul>
		</li>
		<li id="toc-Hipocolesterolemia"
			class="sidebar-toc-list-item sidebar-toc-level-2">
			<a class="sidebar-toc-link" href="#Hipocolesterolemia">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">3.2</span>Hipocolesterolemia</div>
			</a>
			
			<ul id="toc-Hipocolesterolemia-sublist" class="sidebar-toc-list">
			</ul>
		</li>
	</ul>
		</li>
		<li id="toc-Fontes_na_dieta"
			class="sidebar-toc-list-item sidebar-toc-level-1 sidebar-toc-list-item-expanded">
			<a class="sidebar-toc-link" href="#Fontes_na_dieta">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">4</span>Fontes na dieta</div>
			</a>
			
			<ul id="toc-Fontes_na_dieta-sublist" class="sidebar-toc-list">
			</ul>
		</li>
		<li id="toc-Cristais_líquidos_colestéricos"
			class="sidebar-toc-list-item sidebar-toc-level-1 sidebar-toc-list-item-expanded">
			<a class="sidebar-toc-link" href="#Cristais_líquidos_colestéricos">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">5</span>Cristais líquidos colestéricos</div>
			</a>
			
			<ul id="toc-Cristais_líquidos_colestéricos-sublist" class="sidebar-toc-list">
			</ul>
		</li>
		<li id="toc-Imagens_adicionais"
			class="sidebar-toc-list-item sidebar-toc-level-1 sidebar-toc-list-item-expanded">
			<a class="sidebar-toc-link" href="#Imagens_adicionais">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">6</span>Imagens adicionais</div>
			</a>
			
			<ul id="toc-Imagens_adicionais-sublist" class="sidebar-toc-list">
			</ul>
		</li>
		<li id="toc-Ver_também"
			class="sidebar-toc-list-item sidebar-toc-level-1 sidebar-toc-list-item-expanded">
			<a class="sidebar-toc-link" href="#Ver_também">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">7</span>Ver também</div>
			</a>
			
			<ul id="toc-Ver_também-sublist" class="sidebar-toc-list">
			</ul>
		</li>
		<li id="toc-Referências"
			class="sidebar-toc-list-item sidebar-toc-level-1 sidebar-toc-list-item-expanded">
			<a class="sidebar-toc-link" href="#Referências">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">8</span>Referências</div>
			</a>
			
			<ul id="toc-Referências-sublist" class="sidebar-toc-list">
			</ul>
		</li>
		<li id="toc-Ligações_externas"
			class="sidebar-toc-list-item sidebar-toc-level-1 sidebar-toc-list-item-expanded">
			<a class="sidebar-toc-link" href="#Ligações_externas">
				<div class="sidebar-toc-text">
				<span class="sidebar-toc-numb">9</span>Ligações externas</div>
			</a>
			
			<ul id="toc-Ligações_externas-sublist" class="sidebar-toc-list">
			</ul>
		</li>
	</ul>
</nav>

</div>
 <!-- close .vector-sidebar-container -->
<div id="siteNotice"><!-- CentralNotice --><div id="mw-dismissablenotice-anonplace"></div><script>(function(){var node=document.getElementById("mw-dismissablenotice-anonplace");if(node){node.outerHTML="\u003Cdiv class=\"mw-dismissable-notice\"\u003E\u003Cdiv class=\"mw-dismissable-notice-close\"\u003E[\u003Ca tabindex=\"0\" role=\"button\"\u003Eocultar\u003C/a\u003E]\u003C/div\u003E\u003Cdiv class=\"mw-dismissable-notice-body\"\u003E\u003Cdiv id=\"localNotice\"\u003E\u003Cdiv class=\"anonnotice\" lang=\"pt\" dir=\"ltr\"\u003E\u003C/div\u003E\u003C/div\u003E\u003C/div\u003E\u003C/div\u003E";}}());</script></div>
<div class="mw-content-container">
<main id="content" class="mw-body" role="main">
	<a id="top"></a>

	<header class="mw-body-header">
	            
            
            <div id="p-lang-btn" class="vector-menu mw-portlet mw-portlet-lang vector-menu-dropdown"  >
            	<input type="checkbox"
            		id="p-lang-btn-checkbox"
            		role="button"
            		aria-haspopup="true"
            		data-event-name="ui.dropdown-p-lang-btn"
            		class="vector-menu-checkbox  mw-interlanguage-selector "
            		aria-label="Vá para um artigo em outro idioma. Disponível em 97 idiomas"
            	/>
            	<label
            		id="p-lang-btn-label"
            		for="p-lang-btn-checkbox"
            		class="vector-menu-heading mw-ui-button mw-ui-quiet mw-ui-progressive mw-portlet-lang-heading-97"
            	>
            		<span class="mw-ui-icon mw-ui-icon-wikimedia-language-progressive mw-ui-icon-wikimedia-wikimedia-language-progressive"></span><span class="vector-menu-heading-label">97 línguas</span>
            	</label>
            	<div class="vector-menu-content">
            		
            		<ul class="vector-menu-content-list"><li class="interlanguage-link interwiki-af mw-list-item"><a href="https://af.wikipedia.org/wiki/Cholesterol" title="Cholesterol — africanês" lang="af" hreflang="af" class="interlanguage-link-target"><span>Afrikaans</span></a></li><li class="interlanguage-link interwiki-am mw-list-item"><a href="https://am.wikipedia.org/wiki/%E1%8A%AE%E1%88%8C%E1%88%B5%E1%89%B5%E1%88%AE%E1%88%8D" title="ኮሌስትሮል — amárico" lang="am" hreflang="am" class="interlanguage-link-target"><span>አማርኛ</span></a></li><li class="interlanguage-link interwiki-ar mw-list-item"><a href="https://ar.wikipedia.org/wiki/%D9%83%D9%88%D9%84%D8%B3%D8%AA%D8%B1%D9%88%D9%84" title="كولسترول — árabe" lang="ar" hreflang="ar" class="interlanguage-link-target"><span>العربية</span></a></li><li class="interlanguage-link interwiki-as mw-list-item"><a href="https://as.wikipedia.org/wiki/%E0%A6%95%E0%A6%B2%E0%A7%87%E0%A6%B7%E0%A7%8D%E0%A6%9F%E0%A7%87%E0%A7%B0%E0%A6%B2" title="কলেষ্টেৰল — assamês" lang="as" hreflang="as" class="interlanguage-link-target"><span>অসমীয়া</span></a></li><li class="interlanguage-link interwiki-ast mw-list-item"><a href="https://ast.wikipedia.org/wiki/Colesterol" title="Colesterol — asturiano" lang="ast" hreflang="ast" class="interlanguage-link-target"><span>Asturianu</span></a></li><li class="interlanguage-link interwiki-az mw-list-item"><a href="https://az.wikipedia.org/wiki/Xolesterin" title="Xolesterin — azerbaijano" lang="az" hreflang="az" class="interlanguage-link-target"><span>Azərbaycanca</span></a></li><li class="interlanguage-link interwiki-azb mw-list-item"><a href="https://azb.wikipedia.org/wiki/%DA%A9%D9%84%D8%B3%D8%AA%D8%B1%D9%88%D9%84" title="کلسترول — South Azerbaijani" lang="azb" hreflang="azb" class="interlanguage-link-target"><span>تۆرکجه</span></a></li><li class="interlanguage-link interwiki-bcl mw-list-item"><a href="https://bcl.wikipedia.org/wiki/Kolesterol" title="Kolesterol — Central Bikol" lang="bcl" hreflang="bcl" class="interlanguage-link-target"><span>Bikol Central</span></a></li><li class="interlanguage-link interwiki-be mw-list-item"><a href="https://be.wikipedia.org/wiki/%D0%A5%D0%B0%D0%BB%D0%B5%D1%81%D1%82%D1%8D%D1%80%D1%8B%D0%BD" title="Халестэрын — bielorrusso" lang="be" hreflang="be" class="interlanguage-link-target"><span>Беларуская</span></a></li><li class="interlanguage-link interwiki-be-x-old mw-list-item"><a href="https://be-tarask.wikipedia.org/wiki/%D0%A5%D0%B0%D0%BB%D0%B5%D1%81%D1%82%D1%8D%D1%80%D1%8B%D0%BD" title="Халестэрын — Belarusian (Taraškievica orthography)" lang="be-tarask" hreflang="be-tarask" class="interlanguage-link-target"><span>Беларуская (тарашкевіца)</span></a></li><li class="interlanguage-link interwiki-bg mw-list-item"><a href="https://bg.wikipedia.org/wiki/%D0%A5%D0%BE%D0%BB%D0%B5%D1%81%D1%82%D0%B5%D1%80%D0%BE%D0%BB" title="Холестерол — búlgaro" lang="bg" hreflang="bg" class="interlanguage-link-target"><span>Български</span></a></li><li class="interlanguage-link interwiki-bn mw-list-item"><a href="https://bn.wikipedia.org/wiki/%E0%A6%95%E0%A7%8B%E0%A6%B2%E0%A7%87%E0%A6%B8%E0%A7%8D%E0%A6%9F%E0%A7%87%E0%A6%B0%E0%A6%B2" title="কোলেস্টেরল — bengalês" lang="bn" hreflang="bn" class="interlanguage-link-target"><span>বাংলা</span></a></li><li class="interlanguage-link interwiki-br mw-list-item"><a href="https://br.wikipedia.org/wiki/Kolesterol" title="Kolesterol — bretão" lang="br" hreflang="br" class="interlanguage-link-target"><span>Brezhoneg</span></a></li><li class="interlanguage-link interwiki-bs mw-list-item"><a href="https://bs.wikipedia.org/wiki/Holesterol" title="Holesterol — bósnio" lang="bs" hreflang="bs" class="interlanguage-link-target"><span>Bosanski</span></a></li><li class="interlanguage-link interwiki-ca mw-list-item"><a href="https://ca.wikipedia.org/wiki/Colesterol" title="Colesterol — catalão" lang="ca" hreflang="ca" class="interlanguage-link-target"><span>Català</span></a></li><li class="interlanguage-link interwiki-ckb mw-list-item"><a href="https://ckb.wikipedia.org/wiki/%DA%A9%DB%86%D9%84%DB%8C%D8%B3%D8%AA%DA%95%DB%86%DA%B5" title="کۆلیستڕۆڵ — curdo central" lang="ckb" hreflang="ckb" class="interlanguage-link-target"><span>کوردی</span></a></li><li class="interlanguage-link interwiki-cs mw-list-item"><a href="https://cs.wikipedia.org/wiki/Cholesterol" title="Cholesterol — checo" lang="cs" hreflang="cs" class="interlanguage-link-target"><span>Čeština</span></a></li><li class="interlanguage-link interwiki-cv mw-list-item"><a href="https://cv.wikipedia.org/wiki/%D0%A5%D0%BE%D0%BB%D0%B5%D1%81%D1%82%D0%B5%D1%80%D0%B8%D0%BD" title="Холестерин — chuvash" lang="cv" hreflang="cv" class="interlanguage-link-target"><span>Чӑвашла</span></a></li><li class="interlanguage-link interwiki-cy mw-list-item"><a href="https://cy.wikipedia.org/wiki/Cholesterol" title="Cholesterol — galês" lang="cy" hreflang="cy" class="interlanguage-link-target"><span>Cymraeg</span></a></li><li class="interlanguage-link interwiki-da mw-list-item"><a href="https://da.wikipedia.org/wiki/Kolesterol" title="Kolesterol — dinamarquês" lang="da" hreflang="da" class="interlanguage-link-target"><span>Dansk</span></a></li><li class="interlanguage-link interwiki-de mw-list-item"><a href="https://de.wikipedia.org/wiki/Cholesterin" title="Cholesterin — alemão" lang="de" hreflang="de" class="interlanguage-link-target"><span>Deutsch</span></a></li><li class="interlanguage-link interwiki-dv mw-list-item"><a href="https://dv.wikipedia.org/wiki/%DE%86%DE%AE%DE%8D%DE%AC%DE%90%DE%B0%DE%93%DE%B0%DE%83%DE%AF%DE%8D%DE%B0" title="ކޮލެސްޓްރޯލް — divehi" lang="dv" hreflang="dv" class="interlanguage-link-target"><span>ދިވެހިބަސް</span></a></li><li class="interlanguage-link interwiki-el mw-list-item"><a href="https://el.wikipedia.org/wiki/%CE%A7%CE%BF%CE%BB%CE%B7%CF%83%CF%84%CE%B5%CF%81%CF%8C%CE%BB%CE%B7" title="Χοληστερόλη — grego" lang="el" hreflang="el" class="interlanguage-link-target"><span>Ελληνικά</span></a></li><li class="interlanguage-link interwiki-en mw-list-item"><a href="https://en.wikipedia.org/wiki/Cholesterol" title="Cholesterol — inglês" lang="en" hreflang="en" class="interlanguage-link-target"><span>English</span></a></li><li class="interlanguage-link interwiki-eo mw-list-item"><a href="https://eo.wikipedia.org/wiki/Kolesterolo" title="Kolesterolo — esperanto" lang="eo" hreflang="eo" class="interlanguage-link-target"><span>Esperanto</span></a></li><li class="interlanguage-link interwiki-es mw-list-item"><a href="https://es.wikipedia.org/wiki/Colesterol" title="Colesterol — espanhol" lang="es" hreflang="es" class="interlanguage-link-target"><span>Español</span></a></li><li class="interlanguage-link interwiki-et mw-list-item"><a href="https://et.wikipedia.org/wiki/Kolesterool" title="Kolesterool — estónio" lang="et" hreflang="et" class="interlanguage-link-target"><span>Eesti</span></a></li><li class="interlanguage-link interwiki-eu mw-list-item"><a href="https://eu.wikipedia.org/wiki/Kolesterol" title="Kolesterol — basco" lang="eu" hreflang="eu" class="interlanguage-link-target"><span>Euskara</span></a></li><li class="interlanguage-link interwiki-fa mw-list-item"><a href="https://fa.wikipedia.org/wiki/%DA%A9%D9%84%D8%B3%D8%AA%D8%B1%D9%88%D9%84" title="کلسترول — persa" lang="fa" hreflang="fa" class="interlanguage-link-target"><span>فارسی</span></a></li><li class="interlanguage-link interwiki-fi mw-list-item"><a href="https://fi.wikipedia.org/wiki/Kolesteroli" title="Kolesteroli — finlandês" lang="fi" hreflang="fi" class="interlanguage-link-target"><span>Suomi</span></a></li><li class="interlanguage-link interwiki-fr mw-list-item"><a href="https://fr.wikipedia.org/wiki/Cholest%C3%A9rol" title="Cholestérol — francês" lang="fr" hreflang="fr" class="interlanguage-link-target"><span>Français</span></a></li><li class="interlanguage-link interwiki-ga mw-list-item"><a href="https://ga.wikipedia.org/wiki/Colaist%C3%A9ar%C3%B3l" title="Colaistéaról — irlandês" lang="ga" hreflang="ga" class="interlanguage-link-target"><span>Gaeilge</span></a></li><li class="interlanguage-link interwiki-gl mw-list-item"><a href="https://gl.wikipedia.org/wiki/Colesterol" title="Colesterol — galego" lang="gl" hreflang="gl" class="interlanguage-link-target"><span>Galego</span></a></li><li class="interlanguage-link interwiki-he mw-list-item"><a href="https://he.wikipedia.org/wiki/%D7%9B%D7%95%D7%9C%D7%A1%D7%98%D7%A8%D7%95%D7%9C" title="כולסטרול — hebraico" lang="he" hreflang="he" class="interlanguage-link-target"><span>עברית</span></a></li><li class="interlanguage-link interwiki-hi mw-list-item"><a href="https://hi.wikipedia.org/wiki/%E0%A4%95%E0%A5%8B%E0%A4%B2%E0%A5%87%E0%A4%B8%E0%A5%8D%E0%A4%9F%E0%A5%87%E0%A4%B0%E0%A5%89%E0%A4%B2" title="कोलेस्टेरॉल — hindi" lang="hi" hreflang="hi" class="interlanguage-link-target"><span>हिन्दी</span></a></li><li class="interlanguage-link interwiki-hr mw-list-item"><a href="https://hr.wikipedia.org/wiki/Kolesterol" title="Kolesterol — croata" lang="hr" hreflang="hr" class="interlanguage-link-target"><span>Hrvatski</span></a></li><li class="interlanguage-link interwiki-hu mw-list-item"><a href="https://hu.wikipedia.org/wiki/Koleszterin" title="Koleszterin — húngaro" lang="hu" hreflang="hu" class="interlanguage-link-target"><span>Magyar</span></a></li><li class="interlanguage-link interwiki-hy mw-list-item"><a href="https://hy.wikipedia.org/wiki/%D4%BD%D5%B8%D5%AC%D5%A5%D5%BD%D5%A9%D5%A5%D6%80%D5%AB%D5%B6" title="Խոլեսթերին — arménio" lang="hy" hreflang="hy" class="interlanguage-link-target"><span>Հայերեն</span></a></li><li class="interlanguage-link interwiki-ia mw-list-item"><a href="https://ia.wikipedia.org/wiki/Cholesterol" title="Cholesterol — interlíngua" lang="ia" hreflang="ia" class="interlanguage-link-target"><span>Interlingua</span></a></li><li class="interlanguage-link interwiki-id mw-list-item"><a href="https://id.wikipedia.org/wiki/Kolesterol" title="Kolesterol — indonésio" lang="id" hreflang="id" class="interlanguage-link-target"><span>Bahasa Indonesia</span></a></li><li class="interlanguage-link interwiki-is mw-list-item"><a href="https://is.wikipedia.org/wiki/K%C3%B3lester%C3%B3l" title="Kólesteról — islandês" lang="is" hreflang="is" class="interlanguage-link-target"><span>Íslenska</span></a></li><li class="interlanguage-link interwiki-it badge-Q17437798 badge-goodarticle mw-list-item" title="artigo bom"><a href="https://it.wikipedia.org/wiki/Colesterolo" title="Colesterolo — italiano" lang="it" hreflang="it" class="interlanguage-link-target"><span>Italiano</span></a></li><li class="interlanguage-link interwiki-ja badge-Q17437796 badge-featuredarticle mw-list-item" title="artigo destacado"><a href="https://ja.wikipedia.org/wiki/%E3%82%B3%E3%83%AC%E3%82%B9%E3%83%86%E3%83%AD%E3%83%BC%E3%83%AB" title="コレステロール — japonês" lang="ja" hreflang="ja" class="interlanguage-link-target"><span>日本語</span></a></li><li class="interlanguage-link interwiki-jv mw-list-item"><a href="https://jv.wikipedia.org/wiki/Kolesterol" title="Kolesterol — javanês" lang="jv" hreflang="jv" class="interlanguage-link-target"><span>Jawa</span></a></li><li class="interlanguage-link interwiki-ka mw-list-item"><a href="https://ka.wikipedia.org/wiki/%E1%83%A5%E1%83%9D%E1%83%9A%E1%83%94%E1%83%A1%E1%83%A2%E1%83%94%E1%83%A0%E1%83%98%E1%83%9C%E1%83%98" title="ქოლესტერინი — georgiano" lang="ka" hreflang="ka" class="interlanguage-link-target"><span>ქართული</span></a></li><li class="interlanguage-link interwiki-kk mw-list-item"><a href="https://kk.wikipedia.org/wiki/%D0%A5%D0%BE%D0%BB%D0%B5%D1%81%D1%82%D0%B5%D1%80%D0%B8%D0%BD" title="Холестерин — cazaque" lang="kk" hreflang="kk" class="interlanguage-link-target"><span>Қазақша</span></a></li><li class="interlanguage-link interwiki-kn mw-list-item"><a href="https://kn.wikipedia.org/wiki/%E0%B2%95%E0%B3%8A%E0%B2%B2%E0%B3%86%E0%B2%B8%E0%B3%8D%E0%B2%9F%E0%B2%B0%E0%B2%BE%E0%B2%B2%E0%B3%8D%E2%80%8C" title="ಕೊಲೆಸ್ಟರಾಲ್‌ — canarim" lang="kn" hreflang="kn" class="interlanguage-link-target"><span>ಕನ್ನಡ</span></a></li><li class="interlanguage-link interwiki-ko mw-list-item"><a href="https://ko.wikipedia.org/wiki/%EC%BD%9C%EB%A0%88%EC%8A%A4%ED%85%8C%EB%A1%A4" title="콜레스테롤 — coreano" lang="ko" hreflang="ko" class="interlanguage-link-target"><span>한국어</span></a></li><li class="interlanguage-link interwiki-ky mw-list-item"><a href="https://ky.wikipedia.org/wiki/%D0%A5%D0%BE%D0%BB%D0%B5%D1%81%D1%82%D0%B5%D1%80%D0%B8%D0%BD" title="Холестерин — quirguiz" lang="ky" hreflang="ky" class="interlanguage-link-target"><span>Кыргызча</span></a></li><li class="interlanguage-link interwiki-la mw-list-item"><a href="https://la.wikipedia.org/wiki/Cholesterolum" title="Cholesterolum — latim" lang="la" hreflang="la" class="interlanguage-link-target"><span>Latina</span></a></li><li class="interlanguage-link interwiki-lo mw-list-item"><a href="https://lo.wikipedia.org/wiki/%E0%BB%82%E0%BA%81%E0%BB%81%E0%BA%A5%E0%BA%AA%E0%BA%B0%E0%BB%80%E0%BA%95%E0%BB%82%E0%BA%A5%E0%BA%99" title="ໂກແລສະເຕໂລນ — laosiano" lang="lo" hreflang="lo" class="interlanguage-link-target"><span>ລາວ</span></a></li><li class="interlanguage-link interwiki-lt mw-list-item"><a href="https://lt.wikipedia.org/wiki/Cholesterolis" title="Cholesterolis — lituano" lang="lt" hreflang="lt" class="interlanguage-link-target"><span>Lietuvių</span></a></li><li class="interlanguage-link interwiki-lv mw-list-item"><a href="https://lv.wikipedia.org/wiki/Holester%C4%ABns" title="Holesterīns — letão" lang="lv" hreflang="lv" class="interlanguage-link-target"><span>Latviešu</span></a></li><li class="interlanguage-link interwiki-mg mw-list-item"><a href="https://mg.wikipedia.org/wiki/K%C3%B4lesterola" title="Kôlesterola — malgaxe" lang="mg" hreflang="mg" class="interlanguage-link-target"><span>Malagasy</span></a></li><li class="interlanguage-link interwiki-mk mw-list-item"><a href="https://mk.wikipedia.org/wiki/%D0%A5%D0%BE%D0%BB%D0%B5%D1%81%D1%82%D0%B5%D1%80%D0%BE%D0%BB" title="Холестерол — macedónio" lang="mk" hreflang="mk" class="interlanguage-link-target"><span>Македонски</span></a></li><li class="interlanguage-link interwiki-ml mw-list-item"><a href="https://ml.wikipedia.org/wiki/%E0%B4%95%E0%B5%8A%E0%B4%B3%E0%B4%B8%E0%B5%8D%E0%B4%9F%E0%B5%8D%E0%B4%B0%E0%B5%8B%E0%B5%BE" title="കൊളസ്ട്രോൾ — malaiala" lang="ml" hreflang="ml" class="interlanguage-link-target"><span>മലയാളം</span></a></li><li class="interlanguage-link interwiki-mr mw-list-item"><a href="https://mr.wikipedia.org/wiki/%E0%A4%95%E0%A5%89%E0%A4%B2%E0%A5%87%E0%A4%B8%E0%A5%8D%E0%A4%9F%E0%A5%87%E0%A4%B0%E0%A5%89%E0%A4%B2" title="कॉलेस्टेरॉल — marata" lang="mr" hreflang="mr" class="interlanguage-link-target"><span>मराठी</span></a></li><li class="interlanguage-link interwiki-ms mw-list-item"><a href="https://ms.wikipedia.org/wiki/Kolesterol" title="Kolesterol — malaio" lang="ms" hreflang="ms" class="interlanguage-link-target"><span>Bahasa Melayu</span></a></li><li class="interlanguage-link interwiki-my mw-list-item"><a href="https://my.wikipedia.org/wiki/%E1%80%80%E1%80%AD%E1%80%AF%E1%80%9C%E1%80%80%E1%80%BA%E1%80%85%E1%80%91%E1%80%9B%E1%80%B1%E1%80%AC" title="ကိုလက်စထရော — birmanês" lang="my" hreflang="my" class="interlanguage-link-target"><span>မြန်မာဘာသာ</span></a></li><li class="interlanguage-link interwiki-ne mw-list-item"><a href="https://ne.wikipedia.org/wiki/%E0%A4%95%E0%A5%8B%E0%A4%B2%E0%A5%87%E0%A4%B8%E0%A5%8D%E0%A4%9F%E0%A5%87%E0%A4%B0%E0%A5%8B%E0%A4%B2" title="कोलेस्टेरोल — nepalês" lang="ne" hreflang="ne" class="interlanguage-link-target"><span>नेपाली</span></a></li><li class="interlanguage-link interwiki-nl mw-list-item"><a href="https://nl.wikipedia.org/wiki/Cholesterol" title="Cholesterol — neerlandês" lang="nl" hreflang="nl" class="interlanguage-link-target"><span>Nederlands</span></a></li><li class="interlanguage-link interwiki-nn mw-list-item"><a href="https://nn.wikipedia.org/wiki/Kolesterol" title="Kolesterol — norueguês nynorsk" lang="nn" hreflang="nn" class="interlanguage-link-target"><span>Norsk nynorsk</span></a></li><li class="interlanguage-link interwiki-no mw-list-item"><a href="https://no.wikipedia.org/wiki/Kolesterol" title="Kolesterol — norueguês bokmål" lang="nb" hreflang="nb" class="interlanguage-link-target"><span>Norsk bokmål</span></a></li><li class="interlanguage-link interwiki-oc mw-list-item"><a href="https://oc.wikipedia.org/wiki/Colester%C3%B2l" title="Colesteròl — occitano" lang="oc" hreflang="oc" class="interlanguage-link-target"><span>Occitan</span></a></li><li class="interlanguage-link interwiki-or mw-list-item"><a href="https://or.wikipedia.org/wiki/%E0%AC%95%E0%AD%8B%E0%AC%B2%E0%AD%87%E0%AC%B7%E0%AD%8D%E0%AC%9F%E0%AD%8D%E0%AC%B0%E0%AC%B2%E0%AD%8D%E2%80%8C" title="କୋଲେଷ୍ଟ୍ରଲ୍‌ — oriá" lang="or" hreflang="or" class="interlanguage-link-target"><span>ଓଡ଼ିଆ</span></a></li><li class="interlanguage-link interwiki-pl mw-list-item"><a href="https://pl.wikipedia.org/wiki/Cholesterol" title="Cholesterol — polaco" lang="pl" hreflang="pl" class="interlanguage-link-target"><span>Polski</span></a></li><li class="interlanguage-link interwiki-rn mw-list-item"><a href="https://rn.wikipedia.org/wiki/Kolesterole" title="Kolesterole — rundi" lang="rn" hreflang="rn" class="interlanguage-link-target"><span>Ikirundi</span></a></li><li class="interlanguage-link interwiki-ro mw-list-item"><a href="https://ro.wikipedia.org/wiki/Colesterol" title="Colesterol — romeno" lang="ro" hreflang="ro" class="interlanguage-link-target"><span>Română</span></a></li><li class="interlanguage-link interwiki-ru mw-list-item"><a href="https://ru.wikipedia.org/wiki/%D0%A5%D0%BE%D0%BB%D0%B5%D1%81%D1%82%D0%B5%D1%80%D0%B8%D0%BD" title="Холестерин — russo" lang="ru" hreflang="ru" class="interlanguage-link-target"><span>Русский</span></a></li><li class="interlanguage-link interwiki-rw mw-list-item"><a href="https://rw.wikipedia.org/wiki/Cholest%C3%A9rol" title="Cholestérol — quiniaruanda" lang="rw" hreflang="rw" class="interlanguage-link-target"><span>Ikinyarwanda</span></a></li><li class="interlanguage-link interwiki-sah mw-list-item"><a href="https://sah.wikipedia.org/wiki/%D0%A5%D0%BE%D0%BB%D0%B5%D1%81%D1%82%D0%B5%D1%80%D0%B8%D0%BD" title="Холестерин — sakha" lang="sah" hreflang="sah" class="interlanguage-link-target"><span>Саха тыла</span></a></li><li class="interlanguage-link interwiki-sh mw-list-item"><a href="https://sh.wikipedia.org/wiki/Holesterol" title="Holesterol — servo-croata" lang="sh" hreflang="sh" class="interlanguage-link-target"><span>Srpskohrvatski / српскохрватски</span></a></li><li class="interlanguage-link interwiki-si mw-list-item"><a href="https://si.wikipedia.org/wiki/%E0%B6%9A%E0%B7%9C%E0%B6%BD%E0%B7%99%E0%B7%83%E0%B7%8A%E0%B6%A7%E0%B6%BB%E0%B7%9D%E0%B6%BD%E0%B7%8A" title="කොලෙස්ටරෝල් — cingalês" lang="si" hreflang="si" class="interlanguage-link-target"><span>සිංහල</span></a></li><li class="interlanguage-link interwiki-simple mw-list-item"><a href="https://simple.wikipedia.org/wiki/Cholesterol" title="Cholesterol — Simple English" lang="en-simple" hreflang="en-simple" class="interlanguage-link-target"><span>Simple English</span></a></li><li class="interlanguage-link interwiki-sk mw-list-item"><a href="https://sk.wikipedia.org/wiki/Cholesterol" title="Cholesterol — eslovaco" lang="sk" hreflang="sk" class="interlanguage-link-target"><span>Slovenčina</span></a></li><li class="interlanguage-link interwiki-sl mw-list-item"><a href="https://sl.wikipedia.org/wiki/Holesterol" title="Holesterol — esloveno" lang="sl" hreflang="sl" class="interlanguage-link-target"><span>Slovenščina</span></a></li><li class="interlanguage-link interwiki-sq mw-list-item"><a href="https://sq.wikipedia.org/wiki/Kolesteroli" title="Kolesteroli — albanês" lang="sq" hreflang="sq" class="interlanguage-link-target"><span>Shqip</span></a></li><li class="interlanguage-link interwiki-sr mw-list-item"><a href="https://sr.wikipedia.org/wiki/%D0%A5%D0%BE%D0%BB%D0%B5%D1%81%D1%82%D0%B5%D1%80%D0%BE%D0%BB" title="Холестерол — sérvio" lang="sr" hreflang="sr" class="interlanguage-link-target"><span>Српски / srpski</span></a></li><li class="interlanguage-link interwiki-su mw-list-item"><a href="https://su.wikipedia.org/wiki/Kolesterol" title="Kolesterol — sundanês" lang="su" hreflang="su" class="interlanguage-link-target"><span>Sunda</span></a></li><li class="interlanguage-link interwiki-sv mw-list-item"><a href="https://sv.wikipedia.org/wiki/Kolesterol" title="Kolesterol — sueco" lang="sv" hreflang="sv" class="interlanguage-link-target"><span>Svenska</span></a></li><li class="interlanguage-link interwiki-sw mw-list-item"><a href="https://sw.wikipedia.org/wiki/Kolesteroli" title="Kolesteroli — suaíli" lang="sw" hreflang="sw" class="interlanguage-link-target"><span>Kiswahili</span></a></li><li class="interlanguage-link interwiki-ta mw-list-item"><a href="https://ta.wikipedia.org/wiki/%E0%AE%95%E0%AF%8A%E0%AE%B2%E0%AE%B8%E0%AF%8D%E0%AE%9F%E0%AE%BF%E0%AE%B0%E0%AE%BE%E0%AE%B2%E0%AF%8D" title="கொலஸ்டிரால் — tâmil" lang="ta" hreflang="ta" class="interlanguage-link-target"><span>தமிழ்</span></a></li><li class="interlanguage-link interwiki-te mw-list-item"><a href="https://te.wikipedia.org/wiki/%E0%B0%95%E0%B1%8A%E0%B0%B2%E0%B1%86%E0%B0%B8%E0%B1%8D%E0%B0%9F%E0%B1%8D%E0%B0%B0%E0%B0%BE%E0%B0%B2%E0%B1%8D" title="కొలెస్ట్రాల్ — telugu" lang="te" hreflang="te" class="interlanguage-link-target"><span>తెలుగు</span></a></li><li class="interlanguage-link interwiki-th mw-list-item"><a href="https://th.wikipedia.org/wiki/%E0%B8%84%E0%B8%AD%E0%B9%80%E0%B8%A5%E0%B8%AA%E0%B9%80%E0%B8%95%E0%B8%AD%E0%B8%A3%E0%B8%AD%E0%B8%A5" title="คอเลสเตอรอล — tailandês" lang="th" hreflang="th" class="interlanguage-link-target"><span>ไทย</span></a></li><li class="interlanguage-link interwiki-tl mw-list-item"><a href="https://tl.wikipedia.org/wiki/Kolesterol" title="Kolesterol — tagalo" lang="tl" hreflang="tl" class="interlanguage-link-target"><span>Tagalog</span></a></li><li class="interlanguage-link interwiki-tr mw-list-item"><a href="https://tr.wikipedia.org/wiki/Kolesterol" title="Kolesterol — turco" lang="tr" hreflang="tr" class="interlanguage-link-target"><span>Türkçe</span></a></li><li class="interlanguage-link interwiki-tt mw-list-item"><a href="https://tt.wikipedia.org/wiki/%D0%A5%D0%BE%D0%BB%D0%B5%D1%81%D1%82%D0%B5%D1%80%D0%B8%D0%BD" title="Холестерин — tatar" lang="tt" hreflang="tt" class="interlanguage-link-target"><span>Татарча/tatarça</span></a></li><li class="interlanguage-link interwiki-uk mw-list-item"><a href="https://uk.wikipedia.org/wiki/%D0%A5%D0%BE%D0%BB%D0%B5%D1%81%D1%82%D0%B5%D1%80%D0%BE%D0%BB" title="Холестерол — ucraniano" lang="uk" hreflang="uk" class="interlanguage-link-target"><span>Українська</span></a></li><li class="interlanguage-link interwiki-ur mw-list-item"><a href="https://ur.wikipedia.org/wiki/%DA%A9%D9%88%D9%84%DB%8C%D8%B3%D9%B9%D8%B1%D8%A7%D9%84" title="کولیسٹرال — urdu" lang="ur" hreflang="ur" class="interlanguage-link-target"><span>اردو</span></a></li><li class="interlanguage-link interwiki-uz mw-list-item"><a href="https://uz.wikipedia.org/wiki/Xolesterin" title="Xolesterin — usbeque" lang="uz" hreflang="uz" class="interlanguage-link-target"><span>Oʻzbekcha/ўзбекча</span></a></li><li class="interlanguage-link interwiki-vi mw-list-item"><a href="https://vi.wikipedia.org/wiki/Cholesterol" title="Cholesterol — vietnamita" lang="vi" hreflang="vi" class="interlanguage-link-target"><span>Tiếng Việt</span></a></li><li class="interlanguage-link interwiki-wa mw-list-item"><a href="https://wa.wikipedia.org/wiki/Colesterol" title="Colesterol — valão" lang="wa" hreflang="wa" class="interlanguage-link-target"><span>Walon</span></a></li><li class="interlanguage-link interwiki-war mw-list-item"><a href="https://war.wikipedia.org/wiki/Kolesterol" title="Kolesterol — waray" lang="war" hreflang="war" class="interlanguage-link-target"><span>Winaray</span></a></li><li class="interlanguage-link interwiki-wuu mw-list-item"><a href="https://wuu.wikipedia.org/wiki/%E8%83%86%E5%9B%BA%E9%86%87" title="胆固醇 — wu" lang="wuu" hreflang="wuu" class="interlanguage-link-target"><span>吴语</span></a></li><li class="interlanguage-link interwiki-zh mw-list-item"><a href="https://zh.wikipedia.org/wiki/%E8%86%BD%E5%9B%BA%E9%86%87" title="膽固醇 — chinês" lang="zh" hreflang="zh" class="interlanguage-link-target"><span>中文</span></a></li><li class="interlanguage-link interwiki-zh-min-nan mw-list-item"><a href="https://zh-min-nan.wikipedia.org/wiki/Hoeh-i%C3%BB" title="Hoeh-iû — Chinese (Min Nan)" lang="nan" hreflang="nan" class="interlanguage-link-target"><span>Bân-lâm-gú</span></a></li><li class="interlanguage-link interwiki-zh-yue mw-list-item"><a href="https://zh-yue.wikipedia.org/wiki/%E8%86%BD%E5%9B%BA%E9%86%87" title="膽固醇 — Cantonese" lang="yue" hreflang="yue" class="interlanguage-link-target"><span>粵語</span></a></li></ul>
            		<div class="after-portlet after-portlet-lang"><span class="wb-langlinks-edit wb-langlinks-link"><a href="https://www.wikidata.org/wiki/Special:EntityPage/Q43656#sitelinks-wikipedia" title="Editar hiperligações interlínguas" class="wbc-editpage">Editar hiperligações</a></span></div>
            	</div>
            </div>

	
		<label
			id="vector-toc-collapsed-button"
			class="mw-ui-button mw-ui-quiet mw-ui-icon mw-ui-icon-element mw-ui-icon-wikimedia-listBullet mw-checkbox-hack-button"
			for="vector-toc-collapsed-checkbox"
			role="button"
			aria-controls="toc-toggle-list"
			data-event-name="vector.toc-toggle-list"
			tabindex="0"
			title="Table of Contents">
			Toggle the table of contents
		</label>
	    <h1 id="firstHeading" class="firstHeading mw-first-heading">Colesterol</h1>
	</header>
	<nav class="vector-article-toolbar" aria-label="Ferramentas"
		role="navigation">
		<div class="mw-article-toolbar-container">
			<div id="left-navigation">
				

<div id="p-namespaces" class="vector-menu mw-portlet mw-portlet-namespaces vector-menu-tabs"  >
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li id="ca-nstab-main" class="selected mw-list-item"><a href="/wiki/Colesterol" title="Ver a página de conteúdo [c]" accesskey="c"><span>Artigo</span></a></li><li id="ca-talk" class="mw-list-item"><a href="/wiki/Discuss%C3%A3o:Colesterol" rel="discussion" title="Discussão sobre o conteúdo da página [t]" accesskey="t"><span>Discussão</span></a></li></ul>
		
	</div>
</div>

				

<div id="p-variants" class="vector-menu mw-portlet mw-portlet-variants emptyPortlet vector-menu-dropdown-noicon vector-menu-dropdown"  >
	<input type="checkbox"
		id="p-variants-checkbox"
		role="button"
		aria-haspopup="true"
		data-event-name="ui.dropdown-p-variants"
		class="vector-menu-checkbox"
		aria-label="Alterar a variante de idioma"
	/>
	<label
		id="p-variants-label"
		for="p-variants-checkbox"
		class="vector-menu-heading "
	>
		<span class="vector-menu-heading-label">português</span>
	</label>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"></ul>
		
	</div>
</div>

			</div>
			<div id="right-navigation" class="vector-collapsible ">
				

<div id="p-views" class="vector-menu mw-portlet mw-portlet-views vector-menu-tabs"  >
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li id="ca-view" class="selected mw-list-item"><a href="/wiki/Colesterol"><span>Ler</span></a></li><li id="ca-ve-edit" class="mw-list-item"><a href="/w/index.php?title=Colesterol&amp;veaction=edit" title="Editar esta página [v]" accesskey="v"><span>Editar</span></a></li><li id="ca-edit" class="collapsible mw-list-item"><a href="/w/index.php?title=Colesterol&amp;action=edit" title="Editar esta página [e]" accesskey="e"><span>Editar código-fonte</span></a></li><li id="ca-history" class="mw-list-item"><a href="/w/index.php?title=Colesterol&amp;action=history" title="Edições anteriores desta página. [h]" accesskey="h"><span>Ver histórico</span></a></li></ul>
		
	</div>
</div>

				

<div id="p-cactions" class="vector-menu mw-portlet mw-portlet-cactions emptyPortlet vector-menu-dropdown-noicon vector-menu-dropdown vector-has-collapsible-items"  title="Mais opções" >
	<input type="checkbox"
		id="p-cactions-checkbox"
		role="button"
		aria-haspopup="true"
		data-event-name="ui.dropdown-p-cactions"
		class="vector-menu-checkbox"
		
	/>
	<label
		id="p-cactions-label"
		for="p-cactions-checkbox"
		class="vector-menu-heading "
	>
		<span class="vector-menu-heading-label">Mais</span>
	</label>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li id="ca-more-view" class="selected vector-more-collapsible-item mw-list-item"><a href="/wiki/Colesterol"><span>Ler</span></a></li><li id="ca-more-ve-edit" class="vector-more-collapsible-item mw-list-item"><a href="/w/index.php?title=Colesterol&amp;veaction=edit"><span>Editar</span></a></li><li id="ca-more-edit" class="collapsible vector-more-collapsible-item mw-list-item"><a href="/w/index.php?title=Colesterol&amp;action=edit"><span>Editar código-fonte</span></a></li><li id="ca-more-history" class="vector-more-collapsible-item mw-list-item"><a href="/w/index.php?title=Colesterol&amp;action=history"><span>Ver histórico</span></a></li></ul>
		
	</div>
</div>

			</div>
		</div>
	</nav>

	<div id="bodyContent" class="vector-body" data-mw-ve-target-container>
		<div class="mw-body-subheader">
		        <div class="mw-indicators">
        </div>

		    <div id="siteSub" class="noprint">Origem: Wikipédia, a enciclopédia livre.</div>
		</div>		<div id="contentSub"></div>
		<div id="contentSub2"></div>
		
		<div id="mw-content-text" class="mw-body-content mw-content-ltr" lang="pt" dir="ltr"><div class="mw-parser-output"><table class="toccolours" style="width: 300px; float: right; clear: right; margin: 0 0 1em 1em; border-collapse: collapse;">
<tbody><tr>
<th style="background: #F8EABA; text-align: center;" colspan="2">Colesterol<br /><small><a href="/wiki/Wikip%C3%A9dia:Aviso_m%C3%A9dico" title="Wikipédia:Aviso médico">Alerta sobre risco à saúde</a></small>
</th></tr>
<tr>
<td align="center" colspan="2" bgcolor="#ffffff"><a href="/wiki/Ficheiro:Cholesterol.svg" class="image"><img alt="Cholesterol.svg" src="//upload.wikimedia.org/wikipedia/commons/thumb/9/9a/Cholesterol.svg/200px-Cholesterol.svg.png" decoding="async" width="200" height="135" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/9/9a/Cholesterol.svg/300px-Cholesterol.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/9/9a/Cholesterol.svg/400px-Cholesterol.svg.png 2x" data-file-width="440" data-file-height="296" /></a>
</td></tr>






<tr>
<td align="center" colspan="2" bgcolor="#ffffff"><a href="/wiki/Ficheiro:Cholesterol-3d.png" class="image"><img alt="Cholesterol-3d.png" src="//upload.wikimedia.org/wikipedia/commons/thumb/a/ac/Cholesterol-3d.png/200px-Cholesterol-3d.png" decoding="async" width="200" height="141" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/a/ac/Cholesterol-3d.png/300px-Cholesterol-3d.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/a/ac/Cholesterol-3d.png/400px-Cholesterol-3d.png 2x" data-file-width="1422" data-file-height="999" /></a>
</td></tr>







<tr>
<td><a href="/wiki/Nomenclatura_IUPAC" title="Nomenclatura IUPAC">Nome IUPAC</a>
</td>
<td>10,13-dimetil-17-(6-metilheptan-2-yl)-2,3,4,7,8,9,11,12,14,15,16,17-dodecahidro-1H-ciclopenta[a]phenanthren-3-ol
</td></tr>



<tr>
<td>Outros nomes
</td>
<td>(10<i>R</i>,&#x200b;13<i>R</i>)-&#x200b;10,&#x200b;13-&#x200b;dimethyl-&#x200b;17-&#x200b;(6-&#x200b;methylheptan-&#x200b;2-&#x200b;yl)-&#x200b;2,&#x200b;3,&#x200b;4,&#x200b;7,&#x200b;8,&#x200b;9,&#x200b;11,&#x200b;12,&#x200b;14,&#x200b;15,&#x200b;16,&#x200b;17-&#x200b;dodecahydro-&#x200b;1<i>H</i>-&#x200b;cyclopenta&#x200b;[<i>a</i>]phenanthren-&#x200b;3-&#x200b;ol
</td></tr>













































<tr>
<th style="background: #F8EABA; text-align: center;" colspan="2">Identificadores
</th></tr>


<tr>
<td bgcolor=""><a href="/wiki/Registro_CAS" class="mw-redirect" title="Registro CAS">Número CAS</a>
</td>
<td bgcolor=""><span class="reflink plainlinksneverexpand"><a rel="nofollow" class="external text" href="http://www.nlm.nih.gov/cgi/mesh/2009/MB_cgi?term=57-88-5&amp;rn=1">57-88-5</a></span>
</td></tr>

<tr>
<td><a href="/wiki/PubChem" title="PubChem">PubChem</a>
</td>
<td><span class="reflink plainlinksneverexpand"><a rel="nofollow" class="external text" href="http://pubchem.ncbi.nlm.nih.gov/summary/summary.cgi?cid=5997">5997</a></span>
</td></tr>







<tr>
<td><a href="/wiki/ChemSpider" title="ChemSpider">ChemSpider</a>
</td>
<td><span class="reflink plainlinksneverexpand"><a rel="nofollow" class="external text" href="http://www.chemspider.com/Chemical-Structure.5775">5775</a></span>
</td></tr>






 

 


<tr>
<td><a href="/wiki/SMILES" title="SMILES">SMILES</a>
</td>
<td><small><tt><div class="NavFrame collapsed" style="border: none; padding: 0;">
<div class="NavHead" style="font-size: 105%; background: transparent; text-align: left;">&#160;</div>
<ul class="NavContent" style="list-style: none none; margin-left: 0; text-align: left; font-size: 105%; margin-top: 0; margin-bottom: 0; line-height: inherit;"><li style="line-height: inherit; margin: 0">O[C@@H]4C/C3=C/C[C@@H]1[C@H](CC[C@]2([C@H]1CC[C@@H]2[C@H](C)CCCC(C)C)C)[C@@]3(C)CC4</li></ul>
</div></tt></small>
</td></tr>

<tr>
<td><a href="/wiki/Identificador_Qu%C3%ADmico_Internacional" title="Identificador Químico Internacional">InChI</a>
</td>
<td><div style="word-break: break-all;">1/C27H46O/c1-18(2)7-6-8-19(3)23-11-12-24-22-10-9-20-17-21(28)13-15-26<br />(20,4)25(22)14-16-27(23,24)5/h9,18-19,21-25,28H,6-8,10-17H2,1-5H3/t19-,21+,22+,23-,24+,25+,26+,27-/m1/s1</div>
</td></tr>
<tr>
<th style="background: #F8EABA; text-align: center;" colspan="2">Propriedades
</th></tr>
<tr>
<td><a href="/wiki/F%C3%B3rmula_qu%C3%ADmica" title="Fórmula química">Fórmula molecular</a>
</td>
<td>C<sub>27</sub>H<sub>46</sub>O
</td></tr>

<tr>
<td><a href="/wiki/Massa_molar" title="Massa molar">Massa molar</a>
</td>
<td>386.65 g/mol
</td></tr>
<tr>
<td>Aparência
</td>
<td>pó cristalino branco<sup id="cite_ref-MSDS_1-0" class="reference"><a href="#cite_note-MSDS-1"><span>[</span>1<span>]</span></a></sup>
</td></tr>



<tr>
<td><a href="/wiki/Densidade" title="Densidade">Densidade</a>
</td>
<td>1.052 g/cm<sup>3</sup>
</td></tr>

<tr>
<td><a href="/wiki/Ponto_de_fus%C3%A3o" title="Ponto de fusão">Ponto de fusão</a>
</td>
<td>
<p>148–150 °C<sup id="cite_ref-MSDS_1-1" class="reference"><a href="#cite_note-MSDS-1"><span>[</span>1<span>]</span></a></sup> 
</p>
</td></tr>

<tr>
<td><a href="/wiki/Ponto_de_ebuli%C3%A7%C3%A3o" title="Ponto de ebulição">Ponto de ebulição</a>
</td>
<td>
<p>360 °C (decomposto) 
</p>
</td></tr>



<tr>
<td><a href="/wiki/Solubilidade" title="Solubilidade">Solubilidade</a> em <a href="/wiki/%C3%81gua" title="Água">água</a>
</td>
<td>0.095 mg/L (30 °C)
</td></tr>



<tr>
<td><a href="/wiki/Solubilidade" title="Solubilidade">Solubilidade</a>
</td>
<td>solúvel em <a href="/wiki/Acetona" title="Acetona">acetona</a>, <a href="/wiki/Benzeno" title="Benzeno">benzeno</a>, <a href="/wiki/Clorof%C3%B3rmio" title="Clorofórmio">clorofórmio</a>, <a href="/wiki/Etanol" title="Etanol">etanol</a>, <a href="/wiki/%C3%89ter" title="Éter">éter</a>, <a href="/wiki/Hexano" title="Hexano">hexano</a>, <a href="/wiki/Miristato_de_isopropila" title="Miristato de isopropila">miristato de isopropila</a>
</td></tr>





<tr>
<th style="background: #F8EABA; text-align: center;" colspan="2">Compostos relacionados
</th></tr>






<tr>
<td>Compostos relacionados
</td>
<td><a href="/wiki/Lanosterol" title="Lanosterol">Lanosterol</a> (precursor na síntese biológica)
</td></tr>







<tr>
<th style="background: #F8EABA; text-align: center;" colspan="2"><a class="mw-selflink selflink">Página de dados suplementares</a>
</th></tr>
<tr>
<td><a href="/wiki/Colesterol#Estrutura_e_propriedades" title="Colesterol">Estrutura e propriedades</a>
</td>
<td><a href="/wiki/%C3%8Dndice_de_refrac%C3%A7%C3%A3o" class="mw-redirect" title="Índice de refracção"><i>n</i></a>, <a href="/wiki/Constante_diel%C3%A9ctrica" class="mw-redirect" title="Constante dieléctrica">ε<sub>r</sub></a>, etc.
</td></tr>
<tr>
<td><a href="/wiki/Colesterol#Propriedades_termodinâmicas" title="Colesterol">Dados termodinâmicos</a>
</td>
<td>Phase behaviour<br />Solid, liquid, gas
</td></tr>
<tr>
<td><a href="/wiki/Colesterol#Dados_espectrais" title="Colesterol">Dados espectrais</a>
</td>
<td><a href="/wiki/Espectroscopia_UV/vis%C3%ADvel" title="Espectroscopia UV/visível">UV</a>, <a href="/wiki/Espectroscopia_de_infravermelho" title="Espectroscopia de infravermelho">IV</a>, <a href="/wiki/Espectroscopia_de_resson%C3%A2ncia_magn%C3%A9tica_nuclear" class="mw-redirect" title="Espectroscopia de ressonância magnética nuclear">RMN</a>, <a href="/wiki/Espectroscopia_de_massa" class="mw-redirect" title="Espectroscopia de massa">EM</a>
</td></tr>
<tr>
<td style="background: #F8EABA; text-align: center;" colspan="2"><small><b>Exceto onde denotado, os dados referem-se a<br /> materiais sob <a href="/wiki/Condi%C3%A7%C3%B5es_normais_de_temperatura_e_press%C3%A3o" title="Condições normais de temperatura e pressão">condições normais de temperatura e pressão</a></b><br /><b><a href="/wiki/Wikip%C3%A9dia:Informa%C3%A7%C3%A3o_nas_infocaixas_de_qu%C3%ADmica" title="Wikipédia:Informação nas infocaixas de química">Referências e avisos gerais sobre esta caixa</a>.</b><br /><b><a href="/wiki/Wikip%C3%A9dia:Aviso_m%C3%A9dico" title="Wikipédia:Aviso médico">Alerta sobre risco à saúde</a>.</b> </small>
</td></tr>
</tbody></table>
<p><b>Colesterol</b> é um lipídio constituído por um <a href="/wiki/%C3%81lcool" title="Álcool">álcool</a> policíclico de cadeia longa, usualmente denominado um <a href="/wiki/Esteroide" title="Esteroide">esteroide</a>, encontrado nas membranas <a href="/wiki/C%C3%A9lula" title="Célula">celulares</a> e transportado no <a href="/wiki/Plasma" title="Plasma">plasma</a> <a href="/wiki/Sangue" title="Sangue">sanguíneo</a> de todos os <a href="/wiki/Animalia" title="Animalia">animais</a>. É um componente essencial das membranas celulares dos <a href="/wiki/Mam%C3%ADferos" title="Mamíferos">mamíferos</a>. O colesterol é o principal <a href="/wiki/Esterol" title="Esterol">esterol</a> sintetizado pelos animais. Pequenas quantidades são também sintetizadas por outros <a href="/wiki/Eucariotas" class="mw-redirect" title="Eucariotas">eucariotas</a> como fungos, porém alguns procariotas como certas bactérias também são capazes de sintetizá-lo.
</p><p>Os vegetais apresentam o ergosterol. Plantas apresentam um tipo de composto similar chamado de <a href="/wiki/Fitosterol" class="mw-redirect" title="Fitosterol">fitosterol</a>.
</p><p>A maior parte do colesterol presente no corpo é sintetizada pelo próprio organismo, sendo apenas uma pequena parte adquirida pela dieta. Portanto, ao contrário de como se pensava antigamente, o nível de colesterol no sangue não aumenta se não ingerido quantidades adicionais de colesterol através da dieta (a menos, claro, que haja um distúrbio genético). O colesterol é mais abundante nos tecidos que mais sintetizam ou têm membranas densamente agrupadas em maior número, como o <a href="/wiki/F%C3%ADgado" title="Fígado">fígado</a>, <a href="/wiki/Medula_espinhal" title="Medula espinhal">medula espinhal</a>, <a href="/wiki/C%C3%A9rebro" title="Cérebro">cérebro</a> e placas <a href="/wiki/Ateroma" title="Ateroma">ateromatosas</a> (nas <a href="/wiki/Art%C3%A9ria" title="Artéria">artérias</a>). O colesterol tem um papel central em muitos processos <a href="/wiki/Bioqu%C3%ADmica" title="Bioquímica">bioquímicos</a>, mas é mais conhecido pela associação existente entre <a href="/wiki/Doen%C3%A7a_cardiovascular" title="Doença cardiovascular">doenças cardiovasculares</a> e as diversas <a href="/wiki/Lipoprote%C3%ADna" title="Lipoproteína">lipoproteínas</a> que o transportam, e os altos níveis de colesterol no sangue (<a href="/wiki/Hipercolesterolemia" title="Hipercolesterolemia">hipercolesterolemia</a>).
</p><p>O colesterol é insolúvel em água e, consequentemente, insolúvel no sangue. Para ser transportado através da corrente sanguínea ele liga-se a diversos tipos de lipoproteínas, partículas esféricas que tem sua superfície exterior composta principalmente por <a href="/wiki/Prote%C3%ADna" title="Proteína">proteínas</a> <a href="/wiki/Hidrossol%C3%BAvel" class="mw-redirect" title="Hidrossolúvel">hidrossolúveis</a>. Existem vários tipos de lipoproteínas, e elas são classificadas de acordo com a sua densidade. As duas principais lipoproteínas usadas para diagnóstico dos níveis de colesterol são:
</p>
<ul><li><a href="/wiki/Lipoprote%C3%ADna_de_baixa_densidade" title="Lipoproteína de baixa densidade">lipoproteínas de baixa densidade</a> (<i>Low Density Lipoproteins</i> ou LDL): acredita-se que são a classe maléfica ao ser humano, por serem capazes de transportar o colesterol do <a href="/wiki/F%C3%ADgado" title="Fígado">fígado</a> até as <a href="/wiki/C%C3%A9lulas" class="mw-redirect" title="Células">células</a> de vários outros tecidos. Nos últimos anos, o termo (de certa forma impreciso) "colesterol ruim" ou "colesterol mau" tem sido usado para referir ao LDL que, de acordo com a <a href="/w/index.php?title=Hip%C3%B3tese_de_Rudolf_Virchow&amp;action=edit&amp;redlink=1" class="new" title="Hipótese de Rudolf Virchow (página não existe)">hipótese de Rudolf Virchow</a>, acredita-se ter ações danosas (formação de <a href="/wiki/Aterosclerose" title="Aterosclerose">placas ateroscleróticas</a> nos vasos sanguíneos).</li>
<li><a href="/wiki/Lipoprote%C3%ADna_de_alta_densidade" title="Lipoproteína de alta densidade">lipoproteínas de alta densidade</a> (<i>High Density Lipoproteins</i> ou HDL): acredita-se que são capazes de absorver os cristais de colesterol, que começam a ser depositados nas paredes arteriais/veias (retardando o processo arterosclerótico). Tem sido usado o termo "colesterol bom" para referir ao HDL, que se acredita que tem ações benéficas.</li></ul>
<mw:tocplace></mw:tocplace>
<h2><span id="F.C3.B3rmula_e_estrutura_qu.C3.ADmica"></span><span class="mw-headline" id="Fórmula_e_estrutura_química">Fórmula e estrutura química</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=1" class="mw-editsection-visualeditor" title="Editar secção: Fórmula e estrutura química">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=1" title="Editar secção: Fórmula e estrutura química">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h2>
<p>O colesterol é composto por 27 átomos de carbono, tendo como precursores desses carbonos o acetato.  É uma molécula anfipática, ou seja, possui tanto uma porção polar, constituído pelo grupo hidroxila em C-3 como também um grupo apolar, formado pelos quatro anéis do núcleo esteroide e pela cadeia alifática lateral ligada em C-17. O colesterol apresenta estrutura tetracíclica, composto por quatro anéis A, B, C e D.<sup id="cite_ref-2" class="reference"><a href="#cite_note-2"><span>[</span>2<span>]</span></a></sup><sup id="cite_ref-3" class="reference"><a href="#cite_note-3"><span>[</span>3<span>]</span></a></sup>
</p>
<h2><span class="mw-headline" id="Fisiologia">Fisiologia</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=2" class="mw-editsection-visualeditor" title="Editar secção: Fisiologia">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=2" title="Editar secção: Fisiologia">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h2>
<h3><span id="Fun.C3.A7.C3.A3o"></span><span class="mw-headline" id="Função">Função</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=3" class="mw-editsection-visualeditor" title="Editar secção: Função">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=3" title="Editar secção: Função">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h3>
<p>O colesterol é necessário para construir e manter as <a href="/wiki/Membrana_celular" title="Membrana celular">membranas celulares</a>; regula a fluidez da membrana em diversas faixas de <a href="/wiki/Temperatura" title="Temperatura">temperatura</a>. O grupo <a href="/w/index.php?title=Hidroxil&amp;action=edit&amp;redlink=1" class="new" title="Hidroxil (página não existe)">hidroxil</a> presente no colesterol interage com as cabeças <a href="/wiki/Fosfato" title="Fosfato">fosfato</a> da membrana celular, enquanto a maior parte dos <a href="/wiki/Esteroide" title="Esteroide">esteroides</a> e da cadeia de <a href="/wiki/Hidrocarboneto" title="Hidrocarboneto">hidrocarbonetos</a> estão mergulhados no interior da membrana. Algumas pesquisas recentes indicam que o colesterol pode atuar como um <a href="/wiki/Antioxidante" title="Antioxidante">antioxidante</a>.<sup id="cite_ref-Smith1991_4-0" class="reference"><a href="#cite_note-Smith1991-4"><span>[</span>4<span>]</span></a></sup> O colesterol também ajuda na fabricação da <a href="/wiki/B%C3%ADlis" class="mw-redirect" title="Bílis">bílis</a> (que é armazenada na <a href="/wiki/Ves%C3%ADcula_biliar" title="Vesícula biliar">vesícula biliar</a> e ajuda a digerir gorduras), e também é importante para o metabolismo das <a href="/wiki/Vitaminas_lipossol%C3%BAveis" class="mw-redirect" title="Vitaminas lipossolúveis">vitaminas lipossolúveis</a>, incluindo as vitaminas <a href="/wiki/Vitamina_A" title="Vitamina A">A</a>, <a href="/wiki/Vitamina_D" title="Vitamina D">D</a>, <a href="/wiki/Vitamina_E" class="mw-redirect" title="Vitamina E">E</a> e <a href="/wiki/Vitamina_K" title="Vitamina K">K</a>. Ele é o principal precursor para a síntese de vitamina D e de vários <a href="/w/index.php?title=Horm%C3%B4nio_esteroide&amp;action=edit&amp;redlink=1" class="new" title="Hormônio esteroide (página não existe)">hormônios esteroides</a> (que incluem o <a href="/wiki/Cortisol" title="Cortisol">cortisol</a> e a <a href="/wiki/Aldosterona" title="Aldosterona">aldosterona</a> nas <a href="/wiki/Gl%C3%A2ndula_suprarrenal" title="Glândula suprarrenal">glândulas suprarrenais</a>, e os hormônios sexuais <a href="/wiki/Progesterona" title="Progesterona">progesterona</a>, os diversos <a href="/wiki/Estr%C3%B3geno" class="mw-redirect" title="Estrógeno">estrógenos</a>, <a href="/wiki/Testosterona" title="Testosterona">testosterona</a> e derivados).
</p><p>Recentemente, o colesterol também tem sido relacionado a processos de <a href="/wiki/Sinaliza%C3%A7%C3%A3o_celular" title="Sinalização celular">sinalização celular</a>, pela hipótese seria um dos componentes das chamadas "jangadas lipídicas" na <a href="/wiki/Membrana_plasm%C3%A1tica" title="Membrana plasmática">membrana plasmática</a>. Também reduz a permeabilidade da membrana plasmática aos íons de <a href="/wiki/Hidrog%C3%AAnio" class="mw-redirect" title="Hidrogênio">hidrogênio</a> e <a href="/wiki/S%C3%B3dio" title="Sódio">sódio</a>.<sup id="cite_ref-Haines2001_5-0" class="reference"><a href="#cite_note-Haines2001-5"><span>[</span>5<span>]</span></a></sup>
</p>
<h3><span id="S.C3.ADntese_e_ingest.C3.A3o"></span><span class="mw-headline" id="Síntese_e_ingestão">Síntese e ingestão</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=4" class="mw-editsection-visualeditor" title="Editar secção: Síntese e ingestão">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=4" title="Editar secção: Síntese e ingestão">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h3>
<div class="thumb tright"><div class="thumbinner" style="width:282px;"><a href="/wiki/Ficheiro:HMG-CoA_reductase_pathway.png" class="image"><img alt="" src="//upload.wikimedia.org/wikipedia/commons/thumb/3/39/HMG-CoA_reductase_pathway.png/280px-HMG-CoA_reductase_pathway.png" decoding="async" width="280" height="364" class="thumbimage" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/3/39/HMG-CoA_reductase_pathway.png/420px-HMG-CoA_reductase_pathway.png 1.5x, //upload.wikimedia.org/wikipedia/commons/3/39/HMG-CoA_reductase_pathway.png 2x" data-file-width="534" data-file-height="695" /></a>  <div class="thumbcaption"><div class="magnify"><a href="/wiki/Ficheiro:HMG-CoA_reductase_pathway.png" class="internal" title="Ampliar"></a></div>A via metabólica da HMG-CoA redutase</div></div></div>
<p>O colesterol é necessário para o funcionamento normal da membrana plasmática de células de mamíferos, sendo sintetizado no <a href="/wiki/Ret%C3%ADculo_endoplasm%C3%A1tico" title="Retículo endoplasmático">retículo endoplasmático</a> das células ou derivado da dieta, sendo que na segunda fonte é transportado pela via sanguínea pelas lipoproteínas de baixa densidade e é incorporado pelas células através de endocitose mediada por receptores em fossas cobertas de clatrina na membrana plasmática, e então hidrolizados em lisossomas.
</p><p>O colesterol é sintetizado primariamente da <a href="/wiki/Acetil_CoA" class="mw-redirect" title="Acetil CoA">acetil CoA</a> através da cascata da <a href="/wiki/HMG-CoA_redutase" title="HMG-CoA redutase">HMG-CoA redutase</a> em diversas células e tecidos. Cerca de 20 a 25% da produção total diária (~1 <a href="/wiki/Grama" title="Grama">g</a>/<a href="/wiki/Dia" title="Dia">dia</a>) ocorre no <a href="/wiki/F%C3%ADgado" title="Fígado">fígado</a>; outros locais de maior taxa de síntese incluem os <a href="/wiki/Intestino" title="Intestino">intestinos</a>, <a href="/wiki/Gl%C3%A2ndula_adrenal" class="mw-redirect" title="Glândula adrenal">glândulas adrenais</a> e <a href="/wiki/G%C3%B4nadas" class="mw-redirect" title="Gônadas">gônadas</a>. Em uma pessoa de cerca de 68&#160;kg, a quantidade total de colesterol é de 35 g, a produção interna típica diária é de cerca de 1 g e a ingesta é de 200 a 300&#160;mg. Do colesterol liberado ao intestino com a produção de bile, 92-97% é reabsorvido e reciclado via <a href="/wiki/Circula%C3%A7%C3%A3o_entero-hep%C3%A1tica" title="Circulação entero-hepática">circulação entero-hepática</a>.
</p><p>Etapas principais da síntese do colesterol:
</p>
<ol><li>A acetil-CoA se converte em <a href="/wiki/Mevalonato" class="mw-redirect" title="Mevalonato">mevalonato</a>: alta ingestão de <a href="/wiki/Glicose" title="Glicose">glicose</a> leva a <a href="/wiki/Hipercolesterolemia" title="Hipercolesterolemia">hipercolesterolemia</a>.</li>
<li>O mevalonato após reações sucessivas se transforma em <a href="/wiki/Lanosterol" title="Lanosterol">lanosterol</a>.</li>
<li>O lanosterol se converte em colesterol após 21 etapas adicionais. Esse esteroide é sintetizado pelo <a href="/wiki/F%C3%ADgado" title="Fígado">fígado</a>. Através de um processo <a href="/wiki/Homeostase" title="Homeostase">homeostático</a> quanto maior for a ingestão de colesterol, menor será a quantidade sintetizada pelo fígado. Além disto, o colesterol ingerido em quantidades excessivas não consegue ser eliminado em forma de ácidos biliares e o mecanismo de excreção se torna insuficiente.</li></ol>
<p><a href="/wiki/Konrad_Bloch" title="Konrad Bloch">Konrad Bloch</a> e <a href="/wiki/Feodor_Lynen" title="Feodor Lynen">Feodor Lynen</a> dividiram o <a href="/wiki/Nobel_de_Fisiologia_ou_Medicina" title="Nobel de Fisiologia ou Medicina">Nobel de Fisiologia ou Medicina</a> de 1964, pelas suas descobertas sobre os mecanismos de regulação do colesterol e metabolismo de <a href="/wiki/%C3%81cidos_graxos" class="mw-redirect" title="Ácidos graxos">ácidos graxos</a>.
</p>
<h3><span id="Regula.C3.A7.C3.A3o"></span><span class="mw-headline" id="Regulação">Regulação</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=5" class="mw-editsection-visualeditor" title="Editar secção: Regulação">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=5" title="Editar secção: Regulação">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h3>
<p>A biossíntese do colesterol é regulada diretamente pelos níveis presentes do mesmo, apesar dos mecanismos de <a href="/wiki/Homeostase" title="Homeostase">homeostase</a> envolvidos ainda serem apenas parcialmente compreendidos. Uma alta ingestão de colesterol da dieta leva a uma redução global na produção endógena, enquanto que uma ingestão reduzida leva ao efeito oposto. O principal mecanismo regulatório é a sensibilidade do colesterol <a href="/wiki/Intracelular" class="mw-redirect" title="Intracelular">intracelular</a> no <a href="/wiki/Ret%C3%ADculo_endoplasm%C3%A1tico" title="Retículo endoplasmático">retículo endoplasmático</a> pela <a href="/wiki/Prote%C3%ADna" title="Proteína">proteína</a> <a href="/w/index.php?title=SREBP&amp;action=edit&amp;redlink=1" class="new" title="SREBP (página não existe)">de ligação ao elemento de resposta a esterol (SREBP)</a>. Na presença do colesterol, a SREBP se liga a outras duas proteínas: SCAP (SREBP-cleavage activating protein) e <a href="/w/index.php?title=Insig1&amp;action=edit&amp;redlink=1" class="new" title="Insig1 (página não existe)">Insig1</a>. Quando os níveis de colesterol caem, a Insig-1 se dissocia do complexo SREBP-SCAP, permitindo que o complexo migre para o <a href="/wiki/Aparelho_de_Golgi" class="mw-redirect" title="Aparelho de Golgi">aparelho de Golgi</a>, onde a SREBP é clivada pela S1P e S2P (site 1/2 protease), duas enzimas que são ativadas pela SCAP quando os níveis de colesterol estão baixos. A SREBP clivada então migra para o núcleo e age como um <a href="/wiki/Fator_de_transcri%C3%A7%C3%A3o" title="Fator de transcrição">fator de transcrição</a> para se ligar ao elemento regulatório de esterol (<a href="/wiki/SRE" class="mw-redirect" title="SRE">SRE</a>) de diversos genes para estimular sua <a href="/wiki/Transcri%C3%A7%C3%A3o" class="mw-disambig" title="Transcrição">transcrição</a>. Entre os genes transcritos estão o <a href="/w/index.php?title=Receptor_LDL&amp;action=edit&amp;redlink=1" class="new" title="Receptor LDL (página não existe)">receptor LDL</a> e o <a href="/wiki/HMG-CoA_redutase" title="HMG-CoA redutase">HMG-CoA redutase</a>. O primeiro procura por LDL circulante na corrente sanguínea, ao passo que o HMG-CoA redutase leva a uma produção endógena aumentada de colesterol.<sup id="cite_ref-Anderson2003_6-0" class="reference"><a href="#cite_note-Anderson2003-6"><span>[</span>6<span>]</span></a></sup>
</p><p>Uma grande parte deste mecanismo foi esclarecida pelo Dr. <a href="/wiki/Michael_S._Brown" class="mw-redirect" title="Michael S. Brown">Michael S. Brown</a> e Dr. <a href="/wiki/Joseph_L._Goldstein" class="mw-redirect" title="Joseph L. Goldstein">Joseph L. Goldstein</a> nos anos 1970s. Eles receberam o <a href="/wiki/Pr%C3%AAmio_Nobel_de_Fisiologia_ou_Medicina" class="mw-redirect" title="Prêmio Nobel de Fisiologia ou Medicina">Prêmio Nobel de Fisiologia ou Medicina</a> por seu trabalho em 1985.<sup id="cite_ref-Anderson2003_6-1" class="reference"><a href="#cite_note-Anderson2003-6"><span>[</span>6<span>]</span></a></sup>
</p><p>A quantidade média de colesterol no sangue varia com a idade, tipicamente aumentando gradualmente até a pessoa chegar aos sessenta anos de idade. Parece haver variações sazonais nos níveis de colesterol em humanos, aumentando, em média, no inverno.<sup id="cite_ref-Ockene2004_7-0" class="reference"><a href="#cite_note-Ockene2004-7"><span>[</span>7<span>]</span></a></sup>
</p>
<h3><span id="Excre.C3.A7.C3.A3o"></span><span class="mw-headline" id="Excreção">Excreção</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=6" class="mw-editsection-visualeditor" title="Editar secção: Excreção">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=6" title="Editar secção: Excreção">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h3>
<p>O colesterol é excretado do <a href="/wiki/F%C3%ADgado" title="Fígado">fígado</a> na <a href="/wiki/Bile" title="Bile">bile</a> e é reabsorvido nos intestinos. Dentro de certas circunstâncias, quando está mais concentrado, como na <a href="/wiki/Ves%C3%ADcula_biliar" title="Vesícula biliar">vesícula biliar</a>, ele se cristaliza e é um dos principais constituintes da maioria das <a href="/wiki/Pedras_na_ves%C3%ADcula_biliar" class="mw-redirect" title="Pedras na vesícula biliar">pedras na vesícula biliar</a>, embora possam ser formadas, menos frequentemente, pedras de <a href="/wiki/Lecitina" title="Lecitina">lecitina</a> e <a href="/wiki/Bilirrubina" title="Bilirrubina">bilirrubina</a> na vesícula biliar.
</p>
<h3><span class="mw-headline" id="Fluidos_corporais">Fluidos corporais</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=7" class="mw-editsection-visualeditor" title="Editar secção: Fluidos corporais">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=7" title="Editar secção: Fluidos corporais">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h3>
<p>O colesterol é minimamente solúvel em <a href="/wiki/%C3%81gua" title="Água">água</a>; não podendo se dissolver e ser transportado diretamente na corrente sanguínea, que é à base de água. Ao invés, ele é transportado na corrente sanguínea pelas <a href="/wiki/Lipoprote%C3%ADna" title="Lipoproteína">lipoproteínas</a>, que são solúveis em água e carregam o colesterol e <a href="/wiki/Triglicer%C3%ADdeo" class="mw-redirect" title="Triglicerídeo">triglicerídeos</a> internamente. As <a href="/wiki/Apolipoprote%C3%ADna" title="Apolipoproteína">apolipoproteínas</a> que formam a superfície de uma dada partícula de lipoproteína determinam de que células o colesterol será removido e para onde ele será fornecido.
</p><p>As maiores lipoproteínas, que transportam principalmente gorduras da <a href="/wiki/Mucosa" title="Mucosa">mucosa</a> <a href="/wiki/Intestinal" class="mw-redirect" title="Intestinal">intestinal</a> para o fígado, são chamadas de <a href="/wiki/Quilom%C3%ADcron" class="mw-redirect" title="Quilomícron">quilomícrons</a>. Elas carregam principalmente gorduras na forma de triglicerídeos e colesterol. No <a href="/wiki/F%C3%ADgado" title="Fígado">fígado</a>, as partículas de quilomícron liberam triglicerídeos e um pouco de colesterol. O fígado converte os metabólitos dos alimentos não queimados em lipoproteínas de densidade muito baixa (VLDL) e secreta-as no plasma onde são convertidas em partículas de <a href="/wiki/Lipoprote%C3%ADna_de_baixa_densidade" title="Lipoproteína de baixa densidade">lipoproteínas de baixa densidade</a> (LDL) e ácidos graxos não esterificados, que podem afetar outras células do corpo. Em indivíduos saudáveis, as relativamente poucas partículas de LDL são de tamanho grande. Em contraste, números aumentados de partículas de LDL de baixa densidade (sdLDL) são fortemente associados com a presença de doença <a href="/wiki/Ateroma" title="Ateroma">ateromatosa</a> nas artérias. Por esta razão, o LDL é considerado o "colesterol ruim".
</p><p>O programa nacional de educação sobre o colesterol nos Estados Unidos, de 1987, sugere que o níveis de colesterol total no sangue sejam:
</p>
<ul><li>&lt;200&#160;mg/dl colesterol sanguíneo total normal</li>
<li>200–239&#160;mg/dl limite de colesterol total</li>
<li>&gt;240&#160;mg/dl colesterol total alto</li></ul>
<p>As partículas de <a href="/wiki/Lipoprote%C3%ADna_de_alta_densidade" title="Lipoproteína de alta densidade">lipoproteína de alta densidade</a> (HDL) transportam colesterol de volta para o fígado para a excreção, mas variam consideravelmente em sua efetividade em fazer isto. Geralmente é chamada de "colesterol bom", pois ter partículas grandes de HDL em grandes quantidades traz benefícios à saúde. Em contraste, ter pequenas quantidades de partículas grandes de HDL está associado a progressão de doenças com <a href="/wiki/Ateroma" title="Ateroma">ateromas</a> no interior das artérias.
</p>
<h2><span id="Import.C3.A2ncia_cl.C3.ADnica"></span><span class="mw-headline" id="Importância_clínica">Importância clínica</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=8" class="mw-editsection-visualeditor" title="Editar secção: Importância clínica">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=8" title="Editar secção: Importância clínica">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h2>
<h3><span class="mw-headline" id="Hipercolesterolemia">Hipercolesterolemia</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=9" class="mw-editsection-visualeditor" title="Editar secção: Hipercolesterolemia">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=9" title="Editar secção: Hipercolesterolemia">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h3>
<div class="hatnote"><img alt="" src="//upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Magnifying_glass_01.svg/17px-Magnifying_glass_01.svg.png" decoding="async" width="17" height="17" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Magnifying_glass_01.svg/26px-Magnifying_glass_01.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Magnifying_glass_01.svg/34px-Magnifying_glass_01.svg.png 2x" data-file-width="663" data-file-height="659" />Ver artigo&#32;principal: <a href="/wiki/Hipercolesterolemia" title="Hipercolesterolemia">Hipercolesterolemia</a></div>
<p>O termo <a href="/wiki/Hipercolesterolemia" title="Hipercolesterolemia">hipercolesterolemia</a> refere-se a níveis aumentados de colesterol na corrente sanguínea. Condições com elevadas concentrações de partículas LDL oxidadas, especialmente partículas LDL pequenas, estão associadas com a formação de <a href="/wiki/Ateroma" title="Ateroma">ateromas</a> nas paredes das <a href="/wiki/Art%C3%A9ria" title="Artéria">artérias</a>, uma condição conhecida como <a href="/wiki/Aterosclerose" title="Aterosclerose">aterosclerose</a>, que é a principal causa de <a href="/w/index.php?title=Doen%C3%A7a_coronariana_card%C3%ADaca&amp;action=edit&amp;redlink=1" class="new" title="Doença coronariana cardíaca (página não existe)">doença coronariana cardíaca</a> e outras formas de <a href="/wiki/Doen%C3%A7a_card%C3%ADaca" class="mw-redirect" title="Doença cardíaca">doença cardíaca</a>. Em contraste, as partículas de HDL (especialmente HDL grandes) têm sido identificadas como um mecanismo pelo qual o colesterol e mediadores inflamatórios podem ser removidos do ateroma. As taxas aumentadas de HDL estão relacionadas a taxas menores de progressão e até mesmo regressão dos ateromas.
</p><p>Os níveis elevados de frações de lipoproteínas, LDL, IDL e VLDL são considerados <i>aterogênicos</i> (propensos a causas aterosclerose). Os níveis destas frações, ao invés do nível de colesterol total, se relacionam com o aumento e a progressão de aterosclerose. Desta maneira, o nível de colesterol total pode estar dentro dos limites normais, embora composto principalmente de pequenas partículas de LDL e de HDL, o que, sob estas condições, faria com que as taxas de crescimento de ateromas continuariam altas. Em contraste, entretanto, se o número de partículas LDL é baixo (principalmente de partículas grandes) e uma grande porcentagem de partículas de HDL é grande, então as taxas de crescimento de ateromas são geralmente baixas, até mesmo negativas, para qualquer concentração de colesterol total.
</p><p>Estes efeitos são ainda mais complicados pela concentração relativa de <a href="/w/index.php?title=Dimetilarginina_assim%C3%A9trica&amp;action=edit&amp;redlink=1" class="new" title="Dimetilarginina assimétrica (página não existe)">dimetilarginina assimétrica</a> (ADMA) no <a href="/wiki/Endot%C3%A9lio" title="Endotélio">endotélio</a>, já que a ADMA "regula para baixo" (<i>"down-regulation"</i>) a produção de <a href="/wiki/%C3%93xido_n%C3%ADtrico" title="Óxido nítrico">óxido nítrico</a>, um relaxante do endotélio. Consequentemente, níveis altos de ADMA, associados com níveis aumentados de LDL oxidadas proporcionam um fator de risco aumentado à doença cardiovascular.
A <a href="/wiki/Associa%C3%A7%C3%A3o_Americana_do_Cora%C3%A7%C3%A3o" title="Associação Americana do Coração">Associação Americana do Coração</a> relaciona os seguintes níveis de colesterol sanguíneos totais em jejum e o risco para doenças cardíacas:<sup id="cite_ref-AHA_8-0" class="reference"><a href="#cite_note-AHA-8"><span>[</span>8<span>]</span></a></sup>
</p>
<table cellpadding="3" cellspacing="0" border="1" style="border-collapse:collapse">
<tbody><tr>
<td bgcolor="#cccccc"><b>Nível</b> em <a href="/wiki/Miligrama" class="mw-redirect" title="Miligrama">mg</a>/<a href="/wiki/Decilitro" class="mw-redirect" title="Decilitro">dL</a>
</td>
<td bgcolor="#cccccc"><b>Nível</b> em <a href="/wiki/Mol" title="Mol">mmol</a>/<a href="/wiki/Litro" title="Litro">L</a>
</td>
<td bgcolor="#cccccc"><b>Risco de doença cardíaca</b>
</td></tr>
<tr>
<td>&lt;200
</td>
<td>&lt;5,2
</td>
<td>Nível desejável: menor risco de doença cardíaca
</td></tr>
<tr>
<td>200-239
</td>
<td>5,2-6,2
</td>
<td>Limiar de alto risco
</td></tr>
<tr>
<td>&gt;240
</td>
<td>&gt;6,2
</td>
<td>Nível não-desejável: alto risco
</td></tr></tbody></table>
<p>Entretanto, como os métodos atuais de exames determinam o colesterol LDL ("ruim") e o HDL ("bom") separadamente, este modo simplístico de se avaliar o nível de colesterol se tornou obsoleto. O nível desejável de colesterol <a href="/wiki/LDL" class="mw-redirect" title="LDL">LDL</a> é menos do que 100&#160;mg/dL (2,6&#160;mmol/L), embora um novo alvo de &lt;70&#160;mg/dl  pode ser considerado para indivíduos em alto risco baseado em alguns dos testes mencionados acima. O nível ideal de colesterol <a href="/wiki/HDL" class="mw-redirect" title="HDL">HDL</a> é de &gt;60&#160;mg/dl(1.56&#160;mmol/l)  . Uma proporção de colesterol total para o HDL — outra forma útil de medição — de menos de 5:1 acredita-se ser saudável. Como nota, os valores típicos de LDL para criança antes que os estágios iniciais de ateroma comecem a se desenvolver é 35&#160;mg/dL.
</p><p>Os pacientes devem estar conscientes que a maioria dos métodos de examinação para LDL não medem realmente o LDL em seus sangue, uma partícula de tamanho muito menor. Por razões de custo, os valores de LDL têm sido estimados usando-se a fórmula Friedewald: [colesterol total] − [HDL total] − 20% do valor de <a href="/wiki/Triglicer%C3%ADdio" class="mw-redirect" title="Triglicerídio">triglicerídios</a> = LDL estimado. A base disto é que o colesterol total é definido como a soma de HDL, LDL e VLDL. Geralmente somente o colesterol total, o HDL e os triglicerídios são realmente medidos. O VLDL é estimado a quinta parte (1/5) dos triglicerídios. É importante estar em jejum por pelo menos 8-12 horas antes do exame de sangue porque os níveis de triglicerídios variam significativamente com a ingesta de alimentos.
</p><p>Cada vez mais existem evidências clínicas que fortemente sustentam o maior valor de predição dos exames mais sofisticados que medem tanto as concentrações de partículas de LDL e HDL e tamanho, ao contrário dos exames comuns citados acima.
</p>
<h3><span class="mw-headline" id="Hipocolesterolemia">Hipocolesterolemia</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=10" class="mw-editsection-visualeditor" title="Editar secção: Hipocolesterolemia">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=10" title="Editar secção: Hipocolesterolemia">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h3>
<div class="hatnote"><img alt="" src="//upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Magnifying_glass_01.svg/17px-Magnifying_glass_01.svg.png" decoding="async" width="17" height="17" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Magnifying_glass_01.svg/26px-Magnifying_glass_01.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Magnifying_glass_01.svg/34px-Magnifying_glass_01.svg.png 2x" data-file-width="663" data-file-height="659" />Ver artigo&#32;principal: <a href="/wiki/Hipocolesterolemia" title="Hipocolesterolemia">Hipocolesterolemia</a></div>
<p>Os níveis anormalmente baixos de colesterol são chamados de <i><a href="/wiki/Hipocolesterolemia" title="Hipocolesterolemia">hipocolesterolemia</a></i>. As pesquisas sobre as causas desta condição são relativamentes limitadas: enquanto alguns estudos sugerem uma relação com a <a href="/wiki/Depress%C3%A3o_nervosa" class="mw-redirect" title="Depressão nervosa">depressão</a>, <a href="/wiki/C%C3%A2ncer" title="Câncer">câncer</a> e <a href="/wiki/Hemorragia_cerebral" class="mw-redirect" title="Hemorragia cerebral">hemorragia cerebral</a>, ainda não se sabe ao certo se os níveis baixos de colesterol são a causa destas condições ou se são um <a href="/wiki/Epifen%C3%B4meno" class="mw-redirect" title="Epifenômeno">epifenômeno</a>.<a rel="nofollow" class="external autonumber" href="http://www.americanheart.org/presenter.jhtml?identifier=1208">[1]</a>
</p>
<h2><span class="mw-headline" id="Fontes_na_dieta">Fontes na dieta</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=11" class="mw-editsection-visualeditor" title="Editar secção: Fontes na dieta">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=11" title="Editar secção: Fontes na dieta">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h2>
<div class="thumb tright"><div class="thumbinner" style="width:222px;"><a href="/wiki/Ficheiro:Semillas_de_Ch%C3%ADa.jpg" class="image"><img alt="" src="//upload.wikimedia.org/wikipedia/commons/thumb/8/83/Semillas_de_Ch%C3%ADa.jpg/220px-Semillas_de_Ch%C3%ADa.jpg" decoding="async" width="220" height="258" class="thumbimage" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/8/83/Semillas_de_Ch%C3%ADa.jpg/330px-Semillas_de_Ch%C3%ADa.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/8/83/Semillas_de_Ch%C3%ADa.jpg/440px-Semillas_de_Ch%C3%ADa.jpg 2x" data-file-width="2321" data-file-height="2726" /></a>  <div class="thumbcaption"><div class="magnify"><a href="/wiki/Ficheiro:Semillas_de_Ch%C3%ADa.jpg" class="internal" title="Ampliar"></a></div>A semente da <a href="/wiki/Salvia_hispanica" title="Salvia hispanica">chia</a> é uma das mais ricas fontes conhecidas (tanto entre as vegetais quanto entre as animais) do ácido alfalinolênico, um dos ácidos graxos classificados como <a href="/wiki/%C3%94mega_3" title="Ômega 3">ômega 3</a>.</div></div></div>
<p>Na alimentação humana, o colesterol é encontrado nas <a href="/wiki/Gordura" title="Gordura">gorduras</a> animais: todos os alimentos que contêm gorduras animais possuem colesterol, ao passo que os alimentos que não contêm gorduras animais são isentos de colesterol ou possuem quantidades inexpressivas. As principais fontes de colesterol na dieta incluem os <a href="/wiki/Ovo" title="Ovo">ovos</a>, carne de vaca, de galinha e de porco.<sup id="cite_ref-9" class="reference"><a href="#cite_note-9"><span>[</span>9<span>]</span></a></sup><sup id="cite_ref-10" class="reference"><a href="#cite_note-10"><span>[</span>10<span>]</span></a></sup><sup id="cite_ref-11" class="reference"><a href="#cite_note-11"><span>[</span>11<span>]</span></a></sup>
</p><p>Deve-se observar, no entanto, que o colesterol pode ser sintetizado no organismo humano em grandes quantidades, mesmo com uma <a href="/wiki/Vegetarianismo" title="Vegetarianismo">dieta vegetariana</a> ou pobre em colesterol, como resultado de distúrbios no metabolismo. Portanto, elevações pequenas nos níveis de colesterol podem ser inicialmente tratadas apenas com mudança dos hábitos alimentares, mas <a href="/wiki/Hipercolesterolemia" title="Hipercolesterolemia">hipercolesterolemias</a> severas geralmente exigem a associação com tratamento <a href="/wiki/Farmacologia" title="Farmacologia">farmacológico</a>.
</p><p>Produtos vegetais (como <a href="/wiki/Linha%C3%A7a" title="Linhaça">linhaça</a> e <a href="/wiki/Amendoim" title="Amendoim">amendoim</a>) também contêm compostos como o colesterol, os <a href="/wiki/Fitoesterol" title="Fitoesterol">fitoesteróis</a>, que são sugeridos para diminuir os níveis de colesterol no sangue.<sup id="cite_ref-ostlund2003_12-0" class="reference"><a href="#cite_note-ostlund2003-12"><span>[</span>12<span>]</span></a></sup>
</p>
<h2><span id="Cristais_l.C3.ADquidos_colest.C3.A9ricos"></span><span class="mw-headline" id="Cristais_líquidos_colestéricos">Cristais líquidos colestéricos</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=12" class="mw-editsection-visualeditor" title="Editar secção: Cristais líquidos colestéricos">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=12" title="Editar secção: Cristais líquidos colestéricos">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h2>
<p>Alguns derivados do colesterol geram a <i>fase colestérica</i> <a href="/wiki/Cristal_l%C3%ADquido" title="Cristal líquido">cristalina líquida</a>. A fase colestérica é na verdade uma <a href="/w/index.php?title=Fase_nem%C3%A1tica&amp;action=edit&amp;redlink=1" class="new" title="Fase nemática (página não existe)">fase nemática</a> <a href="/wiki/Quiral" class="mw-redirect mw-disambig" title="Quiral">quiral</a>, e muda de cor quando a temperatura é alterada. Dessa maneira, os derivados do colesterol são geralmente usados em <a href="/wiki/Term%C3%B4metro" title="Termômetro">termômetros</a> de cristal líquido e em corantes e tintas sensíveis à temperatura.
</p>
<h2><span class="mw-headline" id="Imagens_adicionais">Imagens adicionais</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=13" class="mw-editsection-visualeditor" title="Editar secção: Imagens adicionais">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=13" title="Editar secção: Imagens adicionais">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h2>
<ul class="gallery mw-gallery-traditional">
		<li class="gallerybox" style="width: 155px"><div style="width: 155px">
			<div class="thumb" style="width: 150px;"><div style="margin:20.5px auto;"><a href="/wiki/Ficheiro:Cholesterol_Spacefill.jpeg" class="image" title="Modelo espacial da molécula de colesterol…"><img alt="Modelo espacial da molécula de colesterol…" src="//upload.wikimedia.org/wikipedia/commons/thumb/2/2c/Cholesterol_Spacefill.jpeg/120px-Cholesterol_Spacefill.jpeg" decoding="async" width="120" height="109" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/2/2c/Cholesterol_Spacefill.jpeg/180px-Cholesterol_Spacefill.jpeg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/2/2c/Cholesterol_Spacefill.jpeg/240px-Cholesterol_Spacefill.jpeg 2x" data-file-width="2689" data-file-height="2439" /></a></div></div>
			<div class="gallerytext">
<p>Modelo espacial da molécula de colesterol…
</p>
			</div>
		</div></li>
</ul>
<h2><span id="Ver_tamb.C3.A9m"></span><span class="mw-headline" id="Ver_também">Ver também</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=14" class="mw-editsection-visualeditor" title="Editar secção: Ver também">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=14" title="Editar secção: Ver também">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h2>
<ul><li><a href="/wiki/Triglicer%C3%ADdios" class="mw-redirect" title="Triglicerídios">Triglicerídios</a></li>
<li><a href="/wiki/Perfil_lip%C3%ADdico" title="Perfil lipídico">Perfil lipídico</a></li></ul>
<h2 style="cursor: help;" title="Esta seção foi configurada para não ser editável diretamente. Edite a página toda ou a seção anterior em vez disso."><span id="Refer.C3.AAncias"></span><span class="mw-headline" id="Referências">Referências</span></h2>
<div class="reflist" style="list-style-type: decimal;"><div class="mw-references-wrap mw-references-columns"><ol class="references">
<li id="cite_note-MSDS-1"><span class="mw-cite-backlink">↑ <sup><i><b><a href="#cite_ref-MSDS_1-0">a</a></b></i></sup> <sup><i><b><a href="#cite_ref-MSDS_1-1">b</a></b></i></sup></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://web.archive.org/web/20070712214801/http://physchem.ox.ac.uk/MSDS/CH/cholesterol.html">«Safety (MSDS) data for cholesterol»</a><span class="reference-accessdate">. Consultado em 20 de outubro de 2007</span>. Arquivado do <a rel="nofollow" class="external text" href="http://physchem.ox.ac.uk/MSDS/CH/cholesterol.html">original</a> em 12 de julho de 2007</cite><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fpt.wikipedia.org%3AColesterol&amp;rft.btitle=Safety+%28MSDS%29+data+for+cholesterol&amp;rft.genre=unknown&amp;rft_id=http%3A%2F%2Fphyschem.ox.ac.uk%2FMSDS%2FCH%2Fcholesterol.html&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-2"><span class="mw-cite-backlink"><a href="#cite_ref-2">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="http://mundoeducacao.bol.uol.com.br/quimica/estrutura-quimica-dos-esteroides.htm">«Estrutura Química dos Esteroides. Química dos Esteroides - Mundo Educação»</a>. <i>Mundo Educação</i><span class="reference-accessdate">. Consultado em 18 de julho de 2016</span></cite><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fpt.wikipedia.org%3AColesterol&amp;rft.atitle=Estrutura+Qu%C3mica+dos+Esteroides.+Qu%C3mica+dos+Esteroides+-+Mundo+Educa%C3%A7%C3%A3o&amp;rft.genre=unknown&amp;rft.jtitle=Mundo+Educa%C3%A7%C3%A3o&amp;rft_id=http%3A%2F%2Fmundoeducacao.bol.uol.com.br%2Fquimica%2Festrutura-quimica-dos-esteroides.htm&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-3"><span class="mw-cite-backlink"><a href="#cite_ref-3">↑</a></span> <span class="reference-text"><cite class="citation book">Nelson, David L.; COX, Michael M (2014). <i>Princípios de Bioquímica de Lehninger</i>. [S.l.]: Artmed. pp.&#160;Porto alegre</cite><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fpt.wikipedia.org%3AColesterol&amp;rft.au=COX%2C+Michael+M&amp;rft.au=Nelson%2C+David+L.&amp;rft.btitle=Princ%C3pios+de+Bioqu%C3mica+de+Lehninger&amp;rft.date=2014&amp;rft.genre=book&amp;rft.pages=Porto+alegre&amp;rft.pub=Artmed&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-Smith1991-4"><span class="mw-cite-backlink"><a href="#cite_ref-Smith1991_4-0">↑</a></span> <span class="reference-text">Smith LL. Another cholesterol hypothesis: cholesterol as antioxidant. <i>Free Radic Biol Med</i> 1991;11:47-61. <a class="external mw-magiclink-pmid" rel="nofollow" href="//www.ncbi.nlm.nih.gov/pubmed/1937129?dopt=Abstract">PMID 1937129</a>.</span>
</li>
<li id="cite_note-Haines2001-5"><span class="mw-cite-backlink"><a href="#cite_ref-Haines2001_5-0">↑</a></span> <span class="reference-text">Haines, TH. Do sterols reduce proton and sodium leaks through lipid bilayers? <i>Prog Lipid Res</i> 2001:40:299 – 324. <a class="external mw-magiclink-pmid" rel="nofollow" href="//www.ncbi.nlm.nih.gov/pubmed/11412894?dopt=Abstract">PMID 11412894</a>.</span>
</li>
<li id="cite_note-Anderson2003-6"><span class="mw-cite-backlink">↑ <sup><i><b><a href="#cite_ref-Anderson2003_6-0">a</a></b></i></sup> <sup><i><b><a href="#cite_ref-Anderson2003_6-1">b</a></b></i></sup></span> <span class="reference-text"><cite class="citation journal">Anderson RG. (2003). «Joe Goldstein and Mike Brown: from cholesterol homeostasis to new paradigms in membrane biology.». <i>Trends Cell Biol</i>. <b>13</b>: 534 – 9. <a href="/wiki/PubMed_Identifier" class="mw-redirect" title="PubMed Identifier">PMID</a>&#160;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/14507481">14507481</a></cite><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fpt.wikipedia.org%3AColesterol&amp;rft.atitle=Joe+Goldstein+and+Mike+Brown%3A+from+cholesterol+homeostasis+to+new+paradigms+in+membrane+biology.&amp;rft.au=Anderson+RG.&amp;rft.date=2003&amp;rft.genre=article&amp;rft.jtitle=Trends+Cell+Biol&amp;rft.pages=534+-+9&amp;rft.volume=13&amp;rft_id=info%3Apmid%2F14507481&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-Ockene2004-7"><span class="mw-cite-backlink"><a href="#cite_ref-Ockene2004_7-0">↑</a></span> <span class="reference-text"><cite class="citation journal">Ockene IS, Chiriboga DE, Stanek EJ 3rd, Harmatz MG, Nicolosi R, Saperia G, Well AD, Freedson P, Merriam PA, Reed G, Ma Y, Matthews CE, Hebert JR. (2004). «Seasonal variation in serum cholesterol levels: treatment implications and possible mechanisms.». <i>Arch Intern Med</i>. <b>164</b>: 863 – 70. <a href="/wiki/PubMed_Identifier" class="mw-redirect" title="PubMed Identifier">PMID</a>&#160;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15111372">15111372</a></cite><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fpt.wikipedia.org%3AColesterol&amp;rft.atitle=Seasonal+variation+in+serum+cholesterol+levels%3A+treatment+implications+and+possible+mechanisms.&amp;rft.au=Ockene+IS%2C+Chiriboga+DE%2C+Stanek+EJ+3rd%2C+Harmatz+MG%2C+Nicolosi+R%2C+Saperia+G%2C+Well+AD%2C+Freedson+P%2C+Merriam+PA%2C+Reed+G%2C+Ma+Y%2C+Matthews+CE%2C+Hebert+JR.&amp;rft.date=2004&amp;rft.genre=article&amp;rft.jtitle=Arch+Intern+Med&amp;rft.pages=863+-+70&amp;rft.volume=164&amp;rft_id=info%3Apmid%2F15111372&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal" class="Z3988"><span style="display:none;">&#160;</span></span><span class="citation-comment" style="display:none; color:#33aa33"> !CS1 manut: Nomes múltiplos: lista de autores (<a href="/wiki/Categoria:!CS1_manut:_Nomes_m%C3%BAltiplos:_lista_de_autores" title="Categoria:!CS1 manut: Nomes múltiplos: lista de autores">link</a>)</span></span>
</li>
<li id="cite_note-AHA-8"><span class="mw-cite-backlink"><a href="#cite_ref-AHA_8-0">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="http://www.americanheart.org/cholesterol/about.jsp">«<span style="padding-left:0.2em;">"</span>About cholesterol" - American Heart Association»</a></cite><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fpt.wikipedia.org%3AColesterol&amp;rft.btitle=%22About+cholesterol%22+-+American+Heart+Association&amp;rft.genre=unknown&amp;rft_id=http%3A%2F%2Fwww.americanheart.org%2Fcholesterol%2Fabout.jsp&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-9"><span class="mw-cite-backlink"><a href="#cite_ref-9">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://www.health.gov/dietaryguidelines/dga2005/report/HTML/table_e18.htm">«Nutrition and Your Health: Dietary Guidelines for Americans. Table E-18. Dietary Sources of Cholesterol Listed in Decreasing Order»</a></cite><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fpt.wikipedia.org%3AColesterol&amp;rft.btitle=Nutrition+and+Your+Health%3A+Dietary+Guidelines+for+Americans.+Table+E-18.+Dietary+Sources+of+Cholesterol+Listed+in+Decreasing+Order&amp;rft.genre=unknown&amp;rft_id=http%3A%2F%2Fwww.health.gov%2Fdietaryguidelines%2Fdga2005%2Freport%2FHTML%2Ftable_e18.htm&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-10"><span class="mw-cite-backlink"><a href="#cite_ref-10">↑</a></span> <span class="reference-text"><cite class="citation web">BRAGAGNOLO, Neura; RODRIGUEZ-AMAYA, Délia (abril de 2002). <a rel="nofollow" class="external text" href="http://www.scielo.br/pdf/%0D/cta/v22n1/a18v22n1.pdf">«Teores de colesterol, lipídios totai e ácidos graxos em cortes de carne suína»</a> <span style="font-size:85%;">(PDF)</span></cite><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fpt.wikipedia.org%3AColesterol&amp;rft.au=RODRIGUEZ-AMAYA%2C+D%C3%A9lia&amp;rft.aufirst=Neura&amp;rft.aulast=BRAGAGNOLO&amp;rft.btitle=Teores+de+colesterol%2C+lip%C3dios+totai+e+%C3%A1cidos+graxos+em+cortes+de+carne+su%C3na&amp;rft.date=2002-04&amp;rft.genre=unknown&amp;rft_id=http%3A%2F%2Fwww.scielo.br%2Fpdf%2F%250D%2Fcta%2Fv22n1%2Fa18v22n1.pdf&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-11"><span class="mw-cite-backlink"><a href="#cite_ref-11">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="http://brasilescola.uol.com.br/quimica/composicao-quimica-colesterol.htm">«Composição Química do Colesterol. Química do Colesterol - Brasil Escola»</a>. <i>Brasil Escola</i><span class="reference-accessdate">. Consultado em 18 de julho de 2016</span></cite><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fpt.wikipedia.org%3AColesterol&amp;rft.atitle=Composi%C3%A7%C3%A3o+Qu%C3mica+do+Colesterol.+Qu%C3mica+do+Colesterol+-+Brasil+Escola&amp;rft.genre=unknown&amp;rft.jtitle=Brasil+Escola&amp;rft_id=http%3A%2F%2Fbrasilescola.uol.com.br%2Fquimica%2Fcomposicao-quimica-colesterol.htm&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-ostlund2003-12"><span class="mw-cite-backlink"><a href="#cite_ref-ostlund2003_12-0">↑</a></span> <span class="reference-text"><cite class="citation journal">Ostlund RE, Racette, SB, and Stenson WF (2003). «Inhibition of cholesterol absorption by phytosterol-replete wheat germ compared with phytosterol-depleted wheat germ». <i>Am J Clin Nutr</i>. <b>77</b> (6): 1385-1589. <a href="/wiki/PubMed_Identifier" class="mw-redirect" title="PubMed Identifier">PMID</a>&#160;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12791614">12791614</a></cite><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fpt.wikipedia.org%3AColesterol&amp;rft.atitle=Inhibition+of+cholesterol+absorption+by+phytosterol-replete+wheat+germ+compared+with+phytosterol-depleted+wheat+germ&amp;rft.au=Ostlund+RE%2C+Racette%2C+SB%2C+and+Stenson+WF&amp;rft.date=2003&amp;rft.genre=article&amp;rft.issue=6&amp;rft.jtitle=Am+J+Clin+Nutr&amp;rft.pages=1385-1589&amp;rft.volume=77&amp;rft_id=info%3Apmid%2F12791614&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal" class="Z3988"><span style="display:none;">&#160;</span></span><span class="citation-comment" style="display:none; color:#33aa33"> !CS1 manut: Nomes múltiplos: lista de autores (<a href="/wiki/Categoria:!CS1_manut:_Nomes_m%C3%BAltiplos:_lista_de_autores" title="Categoria:!CS1 manut: Nomes múltiplos: lista de autores">link</a>)</span></span>
</li>
</ol></div></div>
<h2><span id="Liga.C3.A7.C3.B5es_externas"></span><span class="mw-headline" id="Ligações_externas">Ligações externas</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Colesterol&amp;veaction=edit&amp;section=15" class="mw-editsection-visualeditor" title="Editar secção: Ligações externas">editar</a><span class="mw-editsection-divider"> | </span><a href="/w/index.php?title=Colesterol&amp;action=edit&amp;section=15" title="Editar secção: Ligações externas">editar código-fonte</a><span class="mw-editsection-bracket">]</span></span></h2>
<ul><li><cite class="citation web"><a rel="nofollow" class="external text" href="http://sociedades.cardiol.br/socerj/publico/dica-colesterol.asp">«Cardiol - Colesterol»</a></cite><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fpt.wikipedia.org%3AColesterol&amp;rft.btitle=Cardiol+-+Colesterol&amp;rft.genre=unknown&amp;rft_id=http%3A%2F%2Fsociedades.cardiol.br%2Fsocerj%2Fpublico%2Fdica-colesterol.asp&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></li>
<li><cite class="citation web"><a rel="nofollow" class="external text" href="http://www.sbpc.org.br/comunicacao/sala.imprensa.php?id=4&amp;tp=2">«SBPC/ML - Colesterol»</a></cite><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fpt.wikipedia.org%3AColesterol&amp;rft.btitle=SBPC%2FML+-+Colesterol&amp;rft.genre=unknown&amp;rft_id=http%3A%2F%2Fwww.sbpc.org.br%2Fcomunicacao%2Fsala.imprensa.php%3Fid%3D4%26tp%3D2&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></li>
<li><cite class="citation web"><a rel="nofollow" class="external text" href="http://www.LDLHDL.info">«www.LDLHDL.info»</a> (em inglês)</cite><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fpt.wikipedia.org%3AColesterol&amp;rft.btitle=www.LDLHDL.info&amp;rft.genre=unknown&amp;rft_id=http%3A%2F%2Fwww.LDLHDL.info&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></li></ul>
<div role="navigation" class="navbox" aria-labelledby="Colesterol_e_metabolismo_intermediário_dos_esteroides" style="padding:3px"><table class="nowraplinks collapsible autocollapse navbox-inner" style="border-spacing:0;background:transparent;color:inherit"><tbody><tr><th scope="col" class="navbox-title" colspan="2"><div class="plainlinks hlist navbar mini"><ul><li class="nv-ver"><a href="/wiki/Predefini%C3%A7%C3%A3o:Colesterol_e_esteroides_intermedi%C3%A1rios" title="Predefinição:Colesterol e esteroides intermediários"><abbr title="Ver esta predefinição" style=";;background:none transparent;border:none;-moz-box-shadow:none;-webkit-box-shadow:none;box-shadow:none; padding:0;">v</abbr></a></li><li class="nv-discutir"><a href="/w/index.php?title=Predefini%C3%A7%C3%A3o_Discuss%C3%A3o:Colesterol_e_esteroides_intermedi%C3%A1rios&amp;action=edit&amp;redlink=1" class="new" title="Predefinição Discussão:Colesterol e esteroides intermediários (página não existe)"><abbr title="Discutir esta predefinição" style=";;background:none transparent;border:none;-moz-box-shadow:none;-webkit-box-shadow:none;box-shadow:none; padding:0;">d</abbr></a></li><li class="nv-editar"><a class="external text" href="https://pt.wikipedia.org/w/index.php?title=Predefini%C3%A7%C3%A3o:Colesterol_e_esteroides_intermedi%C3%A1rios&amp;action=edit"><abbr title="Editar esta predefinição" style=";;background:none transparent;border:none;-moz-box-shadow:none;-webkit-box-shadow:none;box-shadow:none; padding:0;">e</abbr></a></li></ul></div><div id="Colesterol_e_metabolismo_intermediário_dos_esteroides" style="font-size:114%;margin:0 4em"><a class="mw-selflink selflink">Colesterol</a> e <a href="/wiki/Metabolismo" title="Metabolismo">metabolismo</a> intermediário dos <a href="/wiki/Esteroide" title="Esteroide">esteroides</a></div></th></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Via_do_mevalonato" title="Via do mevalonato">Via do mevalonato</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"></div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><th scope="row" class="navbox-group" style="width:1%">para <a href="/wiki/HMG-CoA" class="mw-redirect" title="HMG-CoA">HMG-CoA</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Acetilcoenzima_A" title="Acetilcoenzima A">Acetilcoenzima A</a>&#160;&#8226;&#32; <a href="/w/index.php?title=Acetoacetil-CoA&amp;action=edit&amp;redlink=1" class="new" title="Acetoacetil-CoA (página não existe)">Acetoacetil-CoA</a>&#160;&#8226;&#32; <a href="/wiki/HMG-CoA" class="mw-redirect" title="HMG-CoA">HMG-CoA</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Corpos_cet%C3%B4nicos" title="Corpos cetônicos">Corpos cetônicos</a></th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Acetona" title="Acetona">Acetona</a>&#160;&#8226;&#32; <a href="/wiki/%C3%81cido_acetoac%C3%A9tico" title="Ácido acetoacético">Ácido acetoacético</a>&#160;&#8226;&#32; <a href="/wiki/%C3%81cido_beta-hidroxibut%C3%ADrico" title="Ácido beta-hidroxibutírico">Ácido beta-hidroxibutírico</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">para <a href="/wiki/Difosfato_de_dimetilalilo" title="Difosfato de dimetilalilo">DMAPP</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/%C3%81cido_meval%C3%B4nico" class="mw-redirect" title="Ácido mevalônico">Ácido mevalônico</a>&#160;&#8226;&#32; <a href="/w/index.php?title=%C3%81cido_fosfomeval%C3%B4nico&amp;action=edit&amp;redlink=1" class="new" title="Ácido fosfomevalônico (página não existe)">Ácido fosfomevalônico</a>&#160;&#8226;&#32; <a href="/w/index.php?title=%C3%81cido_5-difosfomeval%C3%B4nico&amp;action=edit&amp;redlink=1" class="new" title="Ácido 5-difosfomevalônico (página não existe)">Ácido 5-difosfomevalônico</a>&#160;&#8226;&#32; <a href="/wiki/Isopentenil_pirofosfato" class="mw-redirect" title="Isopentenil pirofosfato">Isopentenil pirofosfato</a>&#160;&#8226;&#32; <a href="/wiki/Dimetilalil_pirofosfato" class="mw-redirect" title="Dimetilalil pirofosfato">Dimetilalil pirofosfato</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Geranil-</th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/w/index.php?title=Geranil_pirofosfato&amp;action=edit&amp;redlink=1" class="new" title="Geranil pirofosfato (página não existe)">Geranil pirofosfato</a>&#160;&#8226;&#32; <a href="/w/index.php?title=Geranilgeranil_pirofosfato&amp;action=edit&amp;redlink=1" class="new" title="Geranilgeranil pirofosfato (página não existe)">Geranilgeranil pirofosfato</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Carotenoide" title="Carotenoide">Carotenoide</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/w/index.php?title=Prefitoeno_difosfato&amp;action=edit&amp;redlink=1" class="new" title="Prefitoeno difosfato (página não existe)">Prefitoeno difosfato</a>&#160;&#8226;&#32; <a href="/w/index.php?title=Fitoeno&amp;action=edit&amp;redlink=1" class="new" title="Fitoeno (página não existe)">Fitoeno</a></div></td></tr></tbody></table><div></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/w/index.php?title=Via_do_n%C3%A3o_mevalonato&amp;action=edit&amp;redlink=1" class="new" title="Via do não mevalonato (página não existe)">Via do não mevalonato</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"></div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><td colspan="2" class="navbox-list navbox-even" style="width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/w/index.php?title=1-Deoxy-D-xylulose_5-phosphate&amp;action=edit&amp;redlink=1" class="new" title="1-Deoxy-D-xylulose 5-phosphate (página não existe)">DOXP</a>&#160;&#8226;&#32;<a href="/w/index.php?title=2-C-methylerythritol_4-phosphate&amp;action=edit&amp;redlink=1" class="new" title="2-C-methylerythritol 4-phosphate (página não existe)">MEP</a>&#160;&#8226;&#32;<a href="/w/index.php?title=4-diphosphocytidyl-2-C-methylerythritol&amp;action=edit&amp;redlink=1" class="new" title="4-diphosphocytidyl-2-C-methylerythritol (página não existe)">CDP-ME</a>&#160;&#8226;&#32;<a href="/w/index.php?title=4-diphosphocytidyl-2-C-methyl-D-erythritol_2-phosphate&amp;action=edit&amp;redlink=1" class="new" title="4-diphosphocytidyl-2-C-methyl-D-erythritol 2-phosphate (página não existe)">CDP-MEP</a>&#160;&#8226;&#32;<a href="/w/index.php?title=2-C-methyl-D-erythritol_2,4-cyclopyrophosphate&amp;action=edit&amp;redlink=1" class="new" title="2-C-methyl-D-erythritol 2,4-cyclopyrophosphate (página não existe)">MEcPP</a>&#160;&#8226;&#32;<a href="/w/index.php?title=(E)-4-hydroxy-3-methyl-but-2-enyl_pyrophosphate&amp;action=edit&amp;redlink=1" class="new" title="(E)-4-hydroxy-3-methyl-but-2-enyl pyrophosphate (página não existe)">HMB-PP</a>&#160;&#8226;&#32;<a href="/w/index.php?title=Isopentenyl_pyrophosphate&amp;action=edit&amp;redlink=1" class="new" title="Isopentenyl pyrophosphate (página não existe)">IPP</a>&#160;&#8226;&#32;<a href="/w/index.php?title=Dimethylallyl_pyrophosphate&amp;action=edit&amp;redlink=1" class="new" title="Dimethylallyl pyrophosphate (página não existe)">DMAPP</a></div></td></tr></tbody></table><div></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Para <a class="mw-selflink selflink">colesterol</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/w/index.php?title=Farnesil_pirofosfato&amp;action=edit&amp;redlink=1" class="new" title="Farnesil pirofosfato (página não existe)">Farnesil pirofosfato</a>&#160;&#8226;&#32; <a href="/w/index.php?title=Escaleno&amp;action=edit&amp;redlink=1" class="new" title="Escaleno (página não existe)">Escaleno</a>&#160;&#8226;&#32; <a href="/w/index.php?title=2,3-oxidoescaleno&amp;action=edit&amp;redlink=1" class="new" title="2,3-oxidoescaleno (página não existe)">2,3-oxidoescaleno</a>&#160;&#8226;&#32; <a href="/wiki/Lanosterol" title="Lanosterol">Lanosterol</a>
<p><a href="/wiki/Lanosterol" title="Lanosterol">Lanosterol</a>&#160;&#8226;&#32; <a href="/w/index.php?title=Latosterol&amp;action=edit&amp;redlink=1" class="new" title="Latosterol (página não existe)">Latosterol</a>&#160;&#8226;&#32; <a href="/w/index.php?title=7-deidrocolesterol&amp;action=edit&amp;redlink=1" class="new" title="7-deidrocolesterol (página não existe)">7-deidrocolesterol</a>&#160;&#8226;&#32; <a class="mw-selflink selflink">Colesterol</a>
</p>
<a href="/wiki/Lanosterol" title="Lanosterol">Lanosterol</a>&#160;&#8226;&#32; <a href="/w/index.php?title=Zimosterol&amp;action=edit&amp;redlink=1" class="new" title="Zimosterol (página não existe)">Zimosterol</a>&#160;&#8226;&#32; <a href="/w/index.php?title=7-deidrodesmosterol&amp;action=edit&amp;redlink=1" class="new" title="7-deidrodesmosterol (página não existe)">7-deidrodesmosterol</a>&#160;&#8226;&#32; <a href="/wiki/Desmosterol" title="Desmosterol">Desmosterol</a>&#160;&#8226;&#32; <a class="mw-selflink selflink">Colesterol</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Esteroide" title="Esteroide">Esteroide</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"></div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><th id="Corticosteroides(C21_pregnane)" scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Corticosteroide" title="Corticosteroide">Corticosteroides</a><br />(<a href="/w/index.php?title=Progestogen&amp;action=edit&amp;redlink=1" class="new" title="Progestogen (página não existe)">C21 pregnane</a>)</th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"></div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Mineralocorticoide" title="Mineralocorticoide">Mineralocorticoides</a></th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Pregnenolona" title="Pregnenolona">Pregnenolona</a>&#160;&#8226;&#32; <a href="/wiki/Progesterona" title="Progesterona">Progesterona</a>&#160;&#8226;&#32; <a href="/w/index.php?title=11-Deoxicorticosterona&amp;action=edit&amp;redlink=1" class="new" title="11-Deoxicorticosterona (página não existe)">11-Deoxicorticosterona</a>&#160;&#8226;&#32; <a href="/wiki/Corticosterona" title="Corticosterona">Corticosterona</a>&#160;&#8226;&#32; <a href="/wiki/Aldosterona" title="Aldosterona">Aldosterona</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Glicocorticoide" title="Glicocorticoide">Glicocorticoides</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Pregnenolona" title="Pregnenolona">Pregnenolona</a>&#160;&#8226;&#32; <a href="/w/index.php?title=17-hidroxipregnenolona&amp;action=edit&amp;redlink=1" class="new" title="17-hidroxipregnenolona (página não existe)">17-hidroxipregnenolona</a>&#160;&#8226;&#32; <a href="/w/index.php?title=17-hidroxiprogesterona&amp;action=edit&amp;redlink=1" class="new" title="17-hidroxiprogesterona (página não existe)">17-hidroxiprogesterona</a>&#160;&#8226;&#32; <a href="/w/index.php?title=11-deoxicortisol&amp;action=edit&amp;redlink=1" class="new" title="11-deoxicortisol (página não existe)">11-deoxicortisol</a>&#160;&#8226;&#32; <a href="/wiki/Cortisol" title="Cortisol">Cortisol</a><br />
<a href="/wiki/Cortisol" title="Cortisol">Cortisol</a>&#160;&#8226;&#32; <a href="/wiki/Cortisona" title="Cortisona">Cortisona</a></div></td></tr></tbody></table><div></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Esteroide_sexual" title="Esteroide sexual">Esteroides sexuais</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"></div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Andr%C3%B3geno" title="Andrógeno">Andrógenos</a><br />(C19 erane)</th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Deidroepiandrosterona" class="mw-redirect" title="Deidroepiandrosterona">DHEA</a>&#160;&#8226;&#32; <a href="/wiki/Androstenediona" title="Androstenediona">Androstenediona</a>/<a href="/w/index.php?title=5-androstenediol&amp;action=edit&amp;redlink=1" class="new" title="5-androstenediol (página não existe)">5-androstenediol</a>&#160;&#8226;&#32; <a href="/wiki/Testosterona" title="Testosterona">Testosterona</a>&#160;&#8226;&#32; <a href="/wiki/Di-hidrotestosterona" title="Di-hidrotestosterona">Di-hidrotestosterona</a><br />
<a href="/w/index.php?title=Sulfato_de_DHEA&amp;action=edit&amp;redlink=1" class="new" title="Sulfato de DHEA (página não existe)">Sulfato de DHEA</a>&#160;&#8226;&#32; <a href="/w/index.php?title=Epitestosterona&amp;action=edit&amp;redlink=1" class="new" title="Epitestosterona (página não existe)">Epitestosterona</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Estrog%C3%AAnio" title="Estrogênio">Estrogênios</a><br />(C18 estrane)</th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Estrona" title="Estrona">Estrona</a>&#160;&#8226;&#32; <a href="/wiki/Estradiol" title="Estradiol">Estradiol</a>&#160;&#8226;&#32; <a href="/wiki/Estriol" title="Estriol">Estriol</a></div></td></tr></tbody></table><div></div></td></tr></tbody></table><div></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Não humanos</th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"></div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Fitosterol" class="mw-redirect" title="Fitosterol">Fitosterois</a></th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/w/index.php?title=Stigmasterol&amp;action=edit&amp;redlink=1" class="new" title="Stigmasterol (página não existe)">Stigmasterol</a>&#160;&#8226;&#32; <a href="/w/index.php?title=Brassicasterol&amp;action=edit&amp;redlink=1" class="new" title="Brassicasterol (página não existe)">Brassicasterol</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Ergosterol" title="Ergosterol">Esgosterois</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Ergosterol" title="Ergosterol">Ergosterol</a>&#160;&#8226;&#32; <a href="/wiki/Ergocalciferol" title="Ergocalciferol">Ergocalciferol</a></div></td></tr></tbody></table><div></div></td></tr><tr><td colspan="2" class="navbox-list navbox-even" style="width:100%;padding:0px"><div style="padding:0em 0.25em"><i>ver também <a href="/wiki/Predefini%C3%A7%C3%A3o:Ester%C3%B3ides_anabolizantes" class="mw-redirect" title="Predefinição:Esteróides anabolizantes">Esteróides anabolizantes</a></i></div></td></tr></tbody></table></div>
<div role="navigation" class="navbox" aria-labelledby="Tipos_de_lípidos" style="padding:3px"><table class="nowraplinks hlist collapsible collapsed navbox-inner" style="border-spacing:0;background:transparent;color:inherit"><tbody><tr><th scope="col" class="navbox-title" colspan="2"><div class="plainlinks hlist navbar mini"><ul><li class="nv-ver"><a href="/wiki/Predefini%C3%A7%C3%A3o:L%C3%ADpidos" title="Predefinição:Lípidos"><abbr title="Ver esta predefinição" style=";;background:none transparent;border:none;-moz-box-shadow:none;-webkit-box-shadow:none;box-shadow:none; padding:0;">v</abbr></a></li><li class="nv-discutir"><a href="/w/index.php?title=Predefini%C3%A7%C3%A3o_Discuss%C3%A3o:L%C3%ADpidos&amp;action=edit&amp;redlink=1" class="new" title="Predefinição Discussão:Lípidos (página não existe)"><abbr title="Discutir esta predefinição" style=";;background:none transparent;border:none;-moz-box-shadow:none;-webkit-box-shadow:none;box-shadow:none; padding:0;">d</abbr></a></li><li class="nv-editar"><a class="external text" href="https://pt.wikipedia.org/w/index.php?title=Predefini%C3%A7%C3%A3o:L%C3%ADpidos&amp;action=edit"><abbr title="Editar esta predefinição" style=";;background:none transparent;border:none;-moz-box-shadow:none;-webkit-box-shadow:none;box-shadow:none; padding:0;">e</abbr></a></li></ul></div><div id="Tipos_de_lípidos" style="font-size:114%;margin:0 4em">Tipos de <a href="/wiki/L%C3%ADpido" title="Lípido">lípidos</a></div></th></tr><tr><th scope="row" class="navbox-group" style="width:1%">Geral</th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Mol%C3%A9cula_anfip%C3%A1tica" title="Molécula anfipática">Anfipático</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Micela" title="Micela">Micela</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Saponifica%C3%A7%C3%A3o" title="Saponificação">Saponificação</a>&#160;<b>&#183;</b>&#32;  <a href="/wiki/Gordura_saturada" title="Gordura saturada">Gordura saturada</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Gordura_insaturada" title="Gordura insaturada">Gordura insaturada</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/%C3%81cido_graxo_essencial" title="Ácido graxo essencial">Ácido gordo essencial</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/%C3%81cido_graxo" title="Ácido graxo">Ácidos gordos</a></th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/%C3%81cido_palm%C3%ADtico" title="Ácido palmítico">Ácido palmítico</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/%C3%81cido_este%C3%A1rico" title="Ácido esteárico">Ácido esteárico</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/%C3%81cido_oleico" title="Ácido oleico">Ácido oleico</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/%C3%81cido_linoleico" title="Ácido linoleico">Ácido linoleico</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/%C3%81cido_alfalinol%C3%AAnico" title="Ácido alfalinolênico">Ácido α-linolénico</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/%C3%81cido_araquid%C3%B4nico" title="Ácido araquidônico">Ácido araquidónico</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/w/index.php?title=Acilglicer%C3%ADdeo&amp;action=edit&amp;redlink=1" class="new" title="Acilglicerídeo (página não existe)">(Acil)glicerídeos</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/w/index.php?title=Monoglicer%C3%ADdeo&amp;action=edit&amp;redlink=1" class="new" title="Monoglicerídeo (página não existe)">Monoglicerídeos</a>&#160;<b>&#183;</b>&#32; <a href="/w/index.php?title=Diglicer%C3%ADdeo&amp;action=edit&amp;redlink=1" class="new" title="Diglicerídeo (página não existe)">Diglicerídeos</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Triacilglicerol" title="Triacilglicerol">Triglicerídeos</a>&#160;<b>&#183;</b>&#32;</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Cera" title="Cera">Ceras</a></th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Cera" title="Cera">Ceras</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/w/index.php?title=Fosfoglic%C3%A9rido&amp;action=edit&amp;redlink=1" class="new" title="Fosfoglicérido (página não existe)">Fosfoglicéridos</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Fosfatidilcolina" title="Fosfatidilcolina">Fosfatidilcolina</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Fosfatidiletanolamina" title="Fosfatidiletanolamina">Fosfatidiletanolamina</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Fosfatidilserina" title="Fosfatidilserina">Fosfatidilserina</a>&#160;<b>&#183;</b>&#32; <a href="/w/index.php?title=Fosfatidilinositol&amp;action=edit&amp;redlink=1" class="new" title="Fosfatidilinositol (página não existe)">Fosfatidilinositol</a>&#160;<b>&#183;</b>&#32; <a href="/w/index.php?title=Fosfatidilglicerol&amp;action=edit&amp;redlink=1" class="new" title="Fosfatidilglicerol (página não existe)">Fosfatidilglicerol</a>&#160;<b>&#183;</b>&#32; <a href="/w/index.php?title=Cardiolipina&amp;action=edit&amp;redlink=1" class="new" title="Cardiolipina (página não existe)">Cardiolipina</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Lecitina" title="Lecitina">Lecitina</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/%C3%81cido_fosfat%C3%ADdico" class="mw-redirect" title="Ácido fosfatídico">Ácido fosfatídico</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/%C3%81cido_lisofosfat%C3%ADdico" title="Ácido lisofosfatídico">Ácido lisofosfatídico</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Esfingol%C3%ADpido" class="mw-redirect" title="Esfingolípido">Esfingolípidos</a></th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Ceramida" title="Ceramida">Ceramidas</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Esfingomielina" title="Esfingomielina">Esfingomielinas</a>&#160;<b>&#183;</b>&#32; <a href="/w/index.php?title=Glicoesfingol%C3%ADpido&amp;action=edit&amp;redlink=1" class="new" title="Glicoesfingolípido (página não existe)">Glicoesfingolípidos</a> &#160;<b>&#183;</b>&#32;</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Glicol%C3%ADpido" class="mw-redirect" title="Glicolípido">Glicolípidos</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Ganglios%C3%ADdio" title="Gangliosídio">Gangliosídios</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Cerebros%C3%ADdio" title="Cerebrosídio">Cerebrosídios</a>&#160;<b>&#183;</b>&#32; <a href="/w/index.php?title=Globos%C3%B3dio&amp;action=edit&amp;redlink=1" class="new" title="Globosódio (página não existe)">Globosódios</a>&#160;<b>&#183;</b>&#32;</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Ester%C3%B3ide" class="mw-redirect" title="Esteróide">Esteróides</a></th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Esterol" title="Esterol">Esterol</a>&#160;<b>&#183;</b>&#32; <a class="mw-selflink selflink">Colesterol</a> &#160;<b>&#183;</b>&#32; <a href="/wiki/Lanosterol" title="Lanosterol">Lanosterol</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Corticosteroide" title="Corticosteroide">Corticoides</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Esteroide_sexual" title="Esteroide sexual">Esteróides sexuais</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Terpeno" class="mw-redirect" title="Terpeno">Terpenos</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Esqualeno" title="Esqualeno">Esqualeno</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Carotenoide" title="Carotenoide">Carotenoides</a> (<a href="/wiki/Caroteno" title="Caroteno">Carotenos</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Xantofila" title="Xantofila">Xantofilas</a>)</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Eicosanoide" title="Eicosanoide">Eicosanoides</a></th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Prostanoide" title="Prostanoide">Prostanoides</a> (<a href="/wiki/Prostaglandina" title="Prostaglandina">Prostaglandinas</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Prostaciclina" title="Prostaciclina">Prostaciclinas</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Tromboxano" title="Tromboxano">Tromboxanos</a>)&#160;<b>&#183;</b>&#32; <a href="/wiki/Leucotrieno" title="Leucotrieno">Leucotrienos</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Componentes</th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a href="/wiki/Esfingosina" title="Esfingosina">Esfingosina</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Glicerol" title="Glicerol">Glicerol</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Isopreno" title="Isopreno">Isopreno</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Colina_(vitamina)" title="Colina (vitamina)">Colina</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Etanolamina" title="Etanolamina">Etanolamina</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Serina" title="Serina">Serina</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Inositol" title="Inositol">Inositol</a>&#160;<b>&#183;</b>&#32; <a href="/wiki/Fosfato" title="Fosfato">Fosfato</a></div></td></tr></tbody></table></div>
<ul class="noprint" style="border-top: solid silver 1px; border-bottom:1px solid silver; padding:3px; background-color: #F9F9F9; text-align: center; margin-top:10px; margin-left: 0; clear: both;"><li style="display: inline;"><span style="white-space: nowrap; margin: auto 1.5em"><span style="margin-right: 0.5em"><a href="/wiki/Ficheiro:Sciences_exactes.svg" title="Portal da química"><img alt="" src="//upload.wikimedia.org/wikipedia/commons/thumb/3/37/Sciences_exactes.svg/25px-Sciences_exactes.svg.png" decoding="async" width="25" height="25" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/3/37/Sciences_exactes.svg/38px-Sciences_exactes.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/3/37/Sciences_exactes.svg/50px-Sciences_exactes.svg.png 2x" data-file-width="256" data-file-height="256" /></a></span> <span style="font-weight:bold"><a href="/wiki/Portal:Qu%C3%ADmica" title="Portal:Química">Portal da química</a></span></span></li>         </ul>
<!-- 
NewPP limit report
Parsed by mw1413
Cached time: 20220724112140
Cache expiry: 1814400
Reduced expiry: false
Complications: []
CPU time usage: 0.355 seconds
Real time usage: 0.450 seconds
Preprocessor visited node count: 4513/1000000
Post‐expand include size: 108927/2097152 bytes
Template argument size: 9679/2097152 bytes
Highest expansion depth: 15/100
Expensive parser function count: 0/500
Unstrip recursion depth: 0/20
Unstrip post‐expand size: 14358/5000000 bytes
Lua time usage: 0.076/10.000 seconds
Lua memory usage: 3968659/52428800 bytes
Number of Wikibase entities loaded: 0/400
-->
<!--
Transclusion expansion time report (%,ms,calls,template)
100.00%  306.417      1 -total
 35.99%  110.275      1 Predefinição:Info/Química
 30.39%   93.129      1 Predefinição:Referências
 19.47%   59.667      8 Predefinição:Navbox
 19.04%   58.355      9 Predefinição:Citar_web
 17.87%   54.753     84 Predefinição:Chembox_entry
 12.15%   37.234      1 Predefinição:Portal3
 11.85%   36.307      1 Predefinição:Chembox_Identifiers
 11.54%   35.364      1 Predefinição:Portal3/Portais
 10.86%   33.288      1 Predefinição:Colesterol_e_esteroides_intermediários
-->

<!-- Saved in parser cache with key ptwiki:pcache:idhash:94201-0!canonical and timestamp 20220724112139 and revision id 61241919.
 -->
</div><noscript><img src="//pt.wikipedia.org/wiki/Special:CentralAutoLogin/start?type=1x1" alt="" title="" width="1" height="1" style="border: none; position: absolute;" /></noscript>
<div class="printfooter">Obtida de "<a dir="ltr" href="https://pt.wikipedia.org/w/index.php?title=Colesterol&amp;oldid=61241919">https://pt.wikipedia.org/w/index.php?title=Colesterol&amp;oldid=61241919</a>"</div></div>
		<div id="catlinks" class="catlinks" data-mw="interface"><div id="mw-normal-catlinks" class="mw-normal-catlinks"><a href="/wiki/Especial:Categorias" title="Especial:Categorias">Categorias</a>: <ul><li><a href="/wiki/Categoria:L%C3%ADpidos" title="Categoria:Lípidos">Lípidos</a></li><li><a href="/wiki/Categoria:Ester%C3%B3is" title="Categoria:Esteróis">Esteróis</a></li></ul></div><div id="mw-hidden-catlinks" class="mw-hidden-catlinks mw-hidden-cats-hidden">Categorias ocultas: <ul><li><a href="/wiki/Categoria:!CS1_manut:_Nomes_m%C3%BAltiplos:_lista_de_autores" title="Categoria:!CS1 manut: Nomes múltiplos: lista de autores">!CS1 manut: Nomes múltiplos: lista de autores</a></li><li><a href="/wiki/Categoria:!P%C3%A1ginas_que_usam_hiperliga%C3%A7%C3%B5es_m%C3%A1gicas_PMID" title="Categoria:!Páginas que usam hiperligações mágicas PMID">!Páginas que usam hiperligações mágicas PMID</a></li><li><a href="/wiki/Categoria:!CS1_ingl%C3%AAs-fontes_em_l%C3%ADngua_(en)" title="Categoria:!CS1 inglês-fontes em língua (en)">!CS1 inglês-fontes em língua (en)</a></li></ul></div></div>
	</div>


</main>

	</div> 
 <!-- close .vector-sidebar-container -->

<div class="mw-workspace-container mw-footer-container">
	
		
<footer id="footer" class="mw-footer" role="contentinfo" >
	<ul id="footer-info">
	<li id="footer-info-lastmod"> Esta página foi editada pela última vez às 23h39min de 25 de maio de 2021.</li>
	<li id="footer-info-copyright">Este texto é disponibilizado nos termos da licença <a href="https://creativecommons.org/licenses/by-sa/3.0/deed.pt">Atribuição-CompartilhaIgual 3.0 Não Adaptada (CC BY-SA 3.0) da Creative Commons</a>;
pode estar sujeito a condições adicionais.
Para mais detalhes, consulte as <a href="https://foundation.wikimedia.org/wiki/Condições_de_Uso">condições de utilização</a>.</li>
</ul>

	<ul id="footer-places">
	<li id="footer-places-privacy"><a href="https://meta.wikimedia.org/wiki/Privacy_policy/pt-br">Política de privacidade</a></li>
	<li id="footer-places-about"><a href="/wiki/Wikip%C3%A9dia:Sobre">Sobre a Wikipédia</a></li>
	<li id="footer-places-disclaimer"><a href="/wiki/Wikip%C3%A9dia:Aviso_geral">Avisos gerais</a></li>
	<li id="footer-places-mobileview"><a href="//pt.m.wikipedia.org/w/index.php?title=Colesterol&amp;mobileaction=toggle_view_mobile" class="noprint stopMobileRedirectToggle">Versão móvel</a></li>
	<li id="footer-places-developers"><a href="https://developer.wikimedia.org">Programadores</a></li>
	<li id="footer-places-statslink"><a href="https://stats.wikimedia.org/#/pt.wikipedia.org">Estatísticas</a></li>
	<li id="footer-places-cookiestatement"><a href="https://foundation.wikimedia.org/wiki/Cookie_statement">Declaração sobre ''cookies''</a></li>
</ul>

	<ul id="footer-icons" class="noprint">
	<li id="footer-copyrightico"><a href="https://wikimediafoundation.org/"><img src="/static/images/footer/wikimedia-button.png" srcset="/static/images/footer/wikimedia-button-1.5x.png 1.5x, /static/images/footer/wikimedia-button-2x.png 2x" width="88" height="31" alt="Wikimedia Foundation" loading="lazy" /></a></li>
	<li id="footer-poweredbyico"><a href="https://www.mediawiki.org/"><img src="/static/images/footer/poweredby_mediawiki_88x31.png" alt="Powered by MediaWiki" srcset="/static/images/footer/poweredby_mediawiki_132x47.png 1.5x, /static/images/footer/poweredby_mediawiki_176x62.png 2x" width="88" height="31" loading="lazy"/></a></li>
</ul>

</footer>

	
</div>
	</div> 
</div> 

<script>(RLQ=window.RLQ||[]).push(function(){mw.config.set({"wgPageParseReport":{"limitreport":{"cputime":"0.355","walltime":"0.450","ppvisitednodes":{"value":4513,"limit":1000000},"postexpandincludesize":{"value":108927,"limit":2097152},"templateargumentsize":{"value":9679,"limit":2097152},"expansiondepth":{"value":15,"limit":100},"expensivefunctioncount":{"value":0,"limit":500},"unstrip-depth":{"value":0,"limit":20},"unstrip-size":{"value":14358,"limit":5000000},"entityaccesscount":{"value":0,"limit":400},"timingprofile":["100.00%  306.417      1 -total"," 35.99%  110.275      1 Predefinição:Info/Química"," 30.39%   93.129      1 Predefinição:Referências"," 19.47%   59.667      8 Predefinição:Navbox"," 19.04%   58.355      9 Predefinição:Citar_web"," 17.87%   54.753     84 Predefinição:Chembox_entry"," 12.15%   37.234      1 Predefinição:Portal3"," 11.85%   36.307      1 Predefinição:Chembox_Identifiers"," 11.54%   35.364      1 Predefinição:Portal3/Portais"," 10.86%   33.288      1 Predefinição:Colesterol_e_esteroides_intermediários"]},"scribunto":{"limitreport-timeusage":{"value":"0.076","limit":"10.000"},"limitreport-memusage":{"value":3968659,"limit":52428800}},"cachereport":{"origin":"mw1413","timestamp":"20220724112140","ttl":1814400,"transientcontent":false}}});});</script>
<script type="application/ld+json">{"@context":"https:\/\/schema.org","@type":"Article","name":"Colesterol","url":"https:\/\/pt.wikipedia.org\/wiki\/Colesterol","sameAs":"http:\/\/www.wikidata.org\/entity\/Q43656","mainEntity":"http:\/\/www.wikidata.org\/entity\/Q43656","author":{"@type":"Organization","name":"Contribuidores dos projetos da Wikimedia"},"publisher":{"@type":"Organization","name":"Funda\u00e7\u00e3o Wikimedia, Inc.","logo":{"@type":"ImageObject","url":"https:\/\/www.wikimedia.org\/static\/images\/wmf-hor-googpub.png"}},"datePublished":"2005-05-19T11:18:19Z","dateModified":"2021-05-25T23:39:03Z","image":"https:\/\/upload.wikimedia.org\/wikipedia\/commons\/9\/9a\/Cholesterol.svg","headline":"composto qu\u00edmico"}</script><script type="application/ld+json">{"@context":"https:\/\/schema.org","@type":"Article","name":"Colesterol","url":"https:\/\/pt.wikipedia.org\/wiki\/Colesterol","sameAs":"http:\/\/www.wikidata.org\/entity\/Q43656","mainEntity":"http:\/\/www.wikidata.org\/entity\/Q43656","author":{"@type":"Organization","name":"Contribuidores dos projetos da Wikimedia"},"publisher":{"@type":"Organization","name":"Funda\u00e7\u00e3o Wikimedia, Inc.","logo":{"@type":"ImageObject","url":"https:\/\/www.wikimedia.org\/static\/images\/wmf-hor-googpub.png"}},"datePublished":"2005-05-19T11:18:19Z","dateModified":"2021-05-25T23:39:03Z","image":"https:\/\/upload.wikimedia.org\/wikipedia\/commons\/9\/9a\/Cholesterol.svg","headline":"composto qu\u00edmico"}</script>
<script>(RLQ=window.RLQ||[]).push(function(){mw.config.set({"wgBackendResponseTime":213,"wgHostname":"mw1369"});});</script>
</body>
</html>

   """
```
