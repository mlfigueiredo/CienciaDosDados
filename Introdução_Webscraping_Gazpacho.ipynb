# Soup

```python
soup.
```

```python
soup.find('a' , {'card': 'card'} )
```

```python
cards = soup.find("a", {"class": "card"})
cards
```

# Strict

```python
cards[0].find("p", {"class": "release__version"})
```

```python
cards[0].find("p", {"class": "release__version"}, partial=False)
```

```python
cards[0].find("p", {"class": "release__version"}, partial=False).text
```

# Attributes

```python
cards[0]
```

```python
cards[0].find('time').attrs
```

```python
cards[0].find('time').attrs['datetime']
```

```python
def parse_card(card):


    version = card.find("p", {"class": "release__version"}, partial=False).text
    timestamp = card.find("time").attrs['datetime']


    
    return {"version": version, 'timestamp': timestamp}
```

```python
parse_card(cards[0])
```

# **Pandas**

```python
[parse_card(x) for x in cards]
```

```python
import pandas as pd

df = pd.DataFrame([parse_card(x) for x in cards])
df
```

```python
df.info()
```

```python
df = pd.DataFrame([parse_card(x) for x in cards]).assign(timestamp=lambda t: pd.to_datetime(t['timestamp']))
df
```
