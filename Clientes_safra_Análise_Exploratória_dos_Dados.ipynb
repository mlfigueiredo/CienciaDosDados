## Análise Estatística

```python
# Estatística Descritiva das variáveis numéricas do Dataset
df.describe()
```

```python
# Frequencia das variáveis categoricas (que possuem ao invés de uma sequencia de números, possuem categorias a, b, c...)
df['renda'].value_counts()
```

```python
df[['endividamento','investimentos']].head()
```

```python
# Criação de Campos 
df['ind_ediv'] = df['endividamento'] / df['investimentos']
df
```

```python
df.corr(method='spearman')
```

```python
import seaborn as sns
ax = sns.heatmap(df.corr(method='spearman'))
```

```python
# Criação de Campos 
df['patrimonio'] = df['investimentos'] - df['endividamento']
df
```

```python
#filtrando registros do dataset com loc
df.loc[df.patrimonio > -1000]
```

```python
#filtrando registros do dataset com loc
df.loc[df.perc_ult_chq_3m > 0.7]
```

```python
df.loc[(df.seguro == 4) | (df.seguro == 2)]
```

```python
df.loc[(df.patrimonio > -2000) & (df.seguro == 1)]
```

```python
#iloc: selecionando pela faixa de ´indice de 2 até 10 (10 é exclusivo)
df.iloc[5:10]
```

```python
df.iloc[5:10, 2:4]
```
