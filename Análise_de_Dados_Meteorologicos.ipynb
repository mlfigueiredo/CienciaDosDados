###### Temperature Data

##### Converter essa parada de Farenheit para Celsios

°F to °C: (°F − 32) × 5/9 = °C

```python
df['AvgTempCelsius']=(df['AvgTemperature']-32) * 5/9
```

```python
df.head()
```

```python
df.drop('AvgTemperature',axis=1,inplace=True)
```

```python
df.head()
```

### Temperatura média anual ao longo do tempo

```python
df.head()
```

```python
plt.figure()
sns.lineplot(x='Year',y='AvgTempCelsius',data=df)
plt.show()
```

```python
df[df['Year'] == 2021].shape
```

### Temperatura média mensal do Brasil ao longo do tempo

```python
df.head()
```

```python
df_mean=    df.groupby     (['Year','Month'])   ['AvgTempCelsius'].mean()
```

```python
df_mean
```

#####  Unstack(Cover the row index to the column axis)

```python
df_mean=df_mean.unstack()
```

```python
df_mean
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(data=df_mean,annot=True,cmap='coolwarm')
plt.show()
```

### Temperatura média de diferentes cidades

```python
df.head()
```

```python
df_city=df.groupby('City')['AvgTempCelsius'].mean()
```

```python
df_city=df_city.to_frame().reset_index()
```

```python
df_city
```

```python
plt.figure()
sns.lineplot(x='City',y='AvgTempCelsius',data=df_city)
plt.show()
```
